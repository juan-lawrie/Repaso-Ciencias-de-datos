## 🏷️ ¿Cuál es el precio más bajo?

Una persona que trabaja en una tienda digital recibe un mensaje de un cliente que dice que vió un producto muy barato la semana pasada, pero que ahora no lo encuentra.

Antes de investigar más, queremos comprobar algo básico: ¿cuál es el precio más bajo en la lista de precios? 🤔

Contamos con una lista de precios en miles de pesos correspondientes a 10 productos: 35, 22, 18, 40, 18, 50, 33, 18, 29, 41.

A simple vista puede parecer fácil decir cuál es el valor más chico...

Pero: ¿cómo podríamos hacer que la compu lo encuentre sola? 🤔

Mirá este código y analizá cada linea. Luego probálo y observá los resultados:

````python
precios = [35, 22, 18, 40, 18, 50, 33, 18, 29, 41]

i = 0
pos_menor = 0
n = len(precios)

while i < n:
    if precios[i] < precios[pos_menor]:
        pos_menor = i
    i = i + 1

print("El precio más bajo es:", precios[pos_menor])
print("Aparece en la posición:", pos_menor)
````

### 🛠️ ¿Qué es un algoritmo?

Un **algoritmo** es simplemente una serie de pasos claros y ordenados para resolver un problema.
Así como una receta te indica cómo preparar una torta, un algoritmo te indica cómo resolver algo con la compu.

En programación, usamos algoritmos todo el tiempo: para buscar datos, ordenarlos, filtrarlos, y más.

En el ejemplo anterior aplicamos un algoritmo de **búsqueda** para encontrar el valor más chico de una lista de números.  

Un buen algoritmo tiene las siguientes características:
- Tiene **instrucciones claras**, sin ambigüedades.
- Se sigue en un **orden lógico** paso a paso.
- Siempre **termina** en algún momento (no se queda dando vueltas infinitamente).

Cuando estamos pensando cómo resolver un problema, muchas veces lo primero que hacemos **no es escribir código**, sino anotar los pasos en un lenguaje más parecido al que hablamos.

Eso se llama **pseudocódigo**: no es código real, pero ayuda a pensar lo que vamos a hacer.

Después, convertimos ese pseudocódigo en un **programa** con un lenguaje que la compu puede entender, como Python 🐍.

En resumen, podés pensarlo así:
1. El **algoritmo** es la receta.
2. El **pseudocódigo** es escribir la receta en tus propias palabras.
3. El **programa** es cuando seguís esa receta en la cocina... ¡con la compu como chef!

## 👀 ¿Está el precio que busco?

Supongamos ahora que queremos saber si un **precio en particular** está en la lista.

Para comprobarlo, aplicaremos un **algoritmo de búsqueda**.

¿Te animás a completar el código? 👇

````python
precios = [35, 22, 18, 40, 18, 50, 33, 18, 29, 41]
precio_buscado = 40

i = 0
pos_buscado = -1
n = len(precios)
encontrado = False

while i < n and not encontrado:
    if precios[i] == precio_buscado:
        pos_buscado = i
        encontrado = True
    i = i + 1

print("El precio buscado es:", precio_buscado)
print("Aparece en la posición:", pos_buscado)
````

### 🔍 Búsqueda lineal

El algoritmo del ejemplo se llama **búsqueda lineal** (o *linear search*, en inglés).

Este algoritmo nos permite **encontrar un valor dentro de una lista** recorriéndola desde el principio hasta el final, **comparando uno por uno**.

En el ejemplo usamos la búsqueda lineal para **encontrar el precio más bajo**, es decir, **el valor mínimo** de una lista.

La idea del algoritmo es simple y la podemos expresar en **pseudocódigo** así:
1. Empezamos con una posición inicial (por ejemplo, `-1` si queremos indicar que no lo encontramos aún).
2. Recorremos la lista, comparando cada elemento con el valor que buscamos.
3. Si lo encontramos, guardamos su posición.

El valor buscado también se suele llamar **clave** (o *key*, en ingles).

El algoritmo se llama **lineal** porque recorre la lista de forma secuencial, es decir, paso a paso y sin saltos.

Además, funciona con cualquier lista, esté ordenada o no. Pero puede ser lento si la lista es muy grande, como ya veremos.

### 🧮 Complejidad algorítmica

Cuando usamos un algoritmo, una de las preguntas que nos hacemos es: **¿cuánto tiempo tarda en ejecutarse?**

Esta pregunta nos introduce en el área de **complejidad algorítmica**, la cual nos ayuda a entender **cómo crece el tiempo de ejecución a medida que aumenta la cantidad de datos**.

En la vida real trabajamos con miles o millones de datos: precios, nombres, imágenes, etc.
Elegir un algoritmo eficiente puede hacer la diferencia entre obtener un resultado en segundos... o en horas.

A veces se habla del **costo computacional** del algoritmo, refiriéndose al tiempo que tarda en ejecutarse. Un algoritmo con **alto costo** es aquel que necesita mucho tiempo (o recursos) cuando crece la cantidad de datos.

En el ejemplo que vimos antes tenemos una lista de precios y queremos saber si un precio (40) aparece en esa lista. La forma más simple de comprobarlo es revisar los valores **uno por uno** desde el principio. Este método, como ya vimos, se llama **búsqueda lineal**.

Pero, ¿cúanto tarda la búsqueda lineal? Bueno, depende del caso:
- **Mejor caso:** el valor buscado está en la primera posición.
- **Peor caso:** el valor buscado está en la última posición o no está en la lista.
- **Caso promedio:** el valor está en una posición intermedia.

En el peor caso, la búsqueda lineal requiere **recorrer toda la lista**, es decir, hacer $n$ pasos si la lista tiene $n$ elementos.

📏 **Notación Big O**

Para hablar de forma general sobre el tiempo que tarda un algoritmo, usamos la **notación Big O**.

Esta notación no indica tiempos exactos, sino cómo **crece el tiempo** con respecto a la **cantidad de datos**, representada con $n$.

En la **búsqueda lineal**, el tiempo de ejecución crece en forma **proporcional** a la cantidad de datos de la lista.

De esta manera, se dice que tiene **complejidad lineal**, porque depende de $n$ y se denota como $O(n)$.

📌 Una forma práctica de estimar la complejidad es ver **cuántos ciclos** tiene el algoritmo. Un solo ciclo que recorre toda la lista probablemente sea $O(n)$.

## 📑 Ordenando precios

Supongamos que necesitamos ordenar los precios de los productos de menor a mayor para tener una vista más organizada en pantalla.

Para hacer esto, necesitamos usar un algoritmo que se encargue de acomodar los elementos en el orden correcto.

Pero, ¿cómo podemos pedirle a la compu que lo haga? 🤔

Mirá este código, lée cada línea y pensá qué hace. Después ejecutalo para ver cómo funciona:

````python
precios = [35, 22, 18, 40, 18, 50, 33, 18, 29, 41]

i = 0
n = len(precios)

while i < n - 1:
    j = i + 1
    indice_menor = i

    while j < n:
        if precios[j] < precios[indice_menor]:
            indice_menor = j
        j = j + 1

    aux = precios[i]
    precios[i] = precios[indice_menor]
    precios[indice_menor] = aux

    i = i + 1

print("Precios ordenados:", precios)
````

### 🔄 Ordenamiento por selección

Un **algoritmo de ordenamiento** sirve para **organizar datos** siguiendo un criterio, por ejemplo, de menor a mayor.

Ordenar una lista nos ayuda analizarla mejor, encontrar tendencias,  detectar valores extremos o *atípicos* (muy altos o muy bajos), etc.

El algoritmo que vimos en el ejemplo se llama **ordenamiento por selección** (o *selection sort*, in inglés) porque en cada paso **selecciona el más elemento más chico** (el mínimo) de la lista y lo coloca en lugar correcto en la lista.

La idea del ordenamiento por selección es parecida a cuanto **se ordenan cartas**: buscás la carta más chica y la ubicás primera. Después buscás la siguiente más chica entre las que quedan y la ubicás en segundo lugar. Y así seguís hasta que todas estén en orden.

En el caso del algoritmo los pasos en **pseudocódigo** serían así:
1. Recorre toda la lista para encontrar el **mínimo**.
2. Lo **intercambia** con el primer elemento.
3. Luego busca el **mínimo** entre los elementos restantes y lo intercambia con el segundo.
4. **Repite** el proceso hasta que la lista esté ordenada.

⏱️ **Complejidad del ordenamiento por selección**


¿Cuánto tarda el algortimo de ordenamiento por selección? Como ya vimos, esto depende del caso:
- **Mejor caso**: la lista ya está ordenada.
- **Peor caso**: la lista ya está ordenada, pero en orden inverso (de mayor a menor).
- **Caso promedio**: los elementos están mezclados sin un patrón u orden claro.

En todos estos casos, el algoritmo siempre necesita **comparar pares de elementos** para determinar si están en el orden correcto.

La diferencia radica en cuántas veces tiene que hacer **intercambios** entre los elementos.

Si lo analizamos contanto el número de comparaciones por iteración, vemos que en la primera iteración se hacen `n-1` comparaciones, en la segunda `n-2` comparaciones, y así hasta 1. El **total de comparaciones** se puede calcular como la suma de:

```python
(n-1) + (n-2) + ... + 2 + 1 = n(n-1)/2

```

El resultado es una expresión **cuadrática**, porque depende de $n^2$.

Entonces decimos que el ordenamiento por selección tiene **complejidad cuadrática**, y se representa como $O(n^2)$.

📌 Esto también podemos verlo observando el **número de ciclos**, como ya dijimos. Este algoritmo utiliza **dos ciclos anidados**, uno que recorre cada posición y otro que busca el mínimo desde esa posición hasta el final. Por eso es $O(n^2)$.

⚠️ **¿Es eficiente?** En cuanto a **eficiencia**, la **complejidad cuadrática** es peor que la **lineal**, porque crece mucho más rapido cuando aumentamos la cantidad de datos $n$.

En general, **evitamos algoritmos cuadráticos** para listas muy grandes, porque pueden tardar mucho.

De todas maneras, el **ordenamiento por selección** es muy útil para aprender, porque es simple de entender y fácil de implementar.

## 🕵️ ¿Hay otra forma de buscar?

Hasta ahora conocimos la forma más directa de buscar un valor en una lista: ir elemento por elemento desde el principio hasta encontrarlo.

Pero si la lista ya está **ordenada**, existe otro método más eficiente para buscar.

A continuación vas a ver un nuevo algoritmo que aprovecha esa ventaja.


Mirá este código con atención y tratá de entender el algoritmo paso a paso. Luego ejecutalo y analizá los resultados:

````python
precios = [18, 18, 18, 22, 29, 33, 35, 40, 41, 50]
precio_buscado = 40

inicio = 0
fin = len(precios) - 1
encontrado = False
pos_buscado = -1

while inicio <= fin and not encontrado:
    medio = (inicio + fin) // 2
    if precios[medio] == precio_buscado:
        pos_buscado = medio
        encontrado = True
    elif precios[medio] < precio_buscado:
        inicio = medio + 1
    else:
        fin = medio - 1

print("El precio buscado es:", precio_buscado)
print("Aparece en la posición:", pos_buscado)
````

### 🧭 Búsqueda binaria

En el ejemplo, conocimos un algoritmo **más eficiente** que la búsqueda lineal para buscar un valor dentro de una lista.

Este algoritmo se llama **búsqueda binaria** y se aplica únicamente cuando la lista está **ordenada**.

En lugar de revisar cada elemento uno por uno, la búsqueda binaria **divide la lista a la mitad en cada paso** y se queda solo con la mitad donde *podría* estar el valor buscado. Esto se repite hasta encontrarlo o hasta que no quede nada por revisar.

📌 Se basa en el principio de **dividir y conquistar**, y se llama *binaria* porque siempre divide en dos partes (*bi = dos*).

Los pasos del algoritmo podrían resumirse así:
1. Define los índices `inicio` y `fin` de la lista.
2. Calcular el índice del **punto medio**: `(inicio + fin) // 2`.
3. Compara el valor del medio con el valor buscado:
    - Si es **igual**: lo encontramos y el algoritmo termina.
    - Si es **menor**: hay que buscar en la **mitad derecha** de la lista (`inicio = medio + 1`).
    - Si es **mayor**: hay que buscar en la **mitad izquierda** de la lista (`fin = medio - 1`).
4. Repite hasta encontrar el valor o hasta que `inicio > fin` (si el valor no está en la lista).

⏱️ **Complejidad de la búsqueda binaria**

La búsqueda binaria tiene una **complejidad logarítmica**, y se expresa como $O(\log n)$.

Esto se debe a que en cada paso el tamaño de la lista se **reduce a la mitad**, pero se hace solo una comparación por paso.

Si empezamos con $n$ elementos, el proceso sigue así:
- Paso 1: quedan $n/2$ elementos.
- Paso 2: quedan $n/4$ elementos.
- Paso 3: quedan $n/8$ elementos.
- ...
- Y así hasta llegar a 1 solo elemento.

Entonce, ¿cuántas veces hay que dividir $n$ entre 2 para obtener 1? La respuesta es $\log_2(n)$.

La diferencia con la búsqueda lineal es notable cuando tenemos muchos datos. Por ejemplo, para una lista de 1.000.000 de elementos, la búsqueda lineal podría necesitar hasta **1.000.000 pasos** en el peor caso, mientras que la búsqueda binaria, sin embargo, solo necesitaría alrededor de **20 pasos** (porque $\log_2(1.000.000) \approx 20$).

💡 **¿Por qué la lista debe estar ordenada?**

La búsqueda binaria **necesita** que la lista esté ordenada para funcionar: el **orden** permite descartar la mitad incorrecta en cada paso.

Si el valor buscado es mayor que el valor del medio, sabemos que el valor puede estar en la mitad derecha.

Si el valor buscado es menor, sabemos que puede estar en la mitad izquierda.

Sin un orden en los elementos, no sería posible hacer estas suposiciones y la búsqueda binaria no funcionaría correctamente.

### 🧮 Complejidad combinada

A veces un algoritmo tiene **varias etapas**, y cada etapa tiene su propia complejidad.

En el ejemplo, el cálculo del desvío estándar tiene **dos partes**:
- Primer `while`: calcula la **media**, recorriendo todos los datos.
- Segundo `while`: calcula la **sumatoria de cuadrados**, también recorriendo todos los datos.

Ambas partes tienen **complejidad lineal**, el decir $O(n)$ y se ejecutan de forma secuencial (no están anidadas).

Entonces, la complejidad total es: $O(n)+O(n)=O(2n)= O(n)$

📈 **¿Qué mide la notación Big O?**

Más que contar **cuántos pasos exactos** hace un algoritmo, nos interesa ver **cómo crece** ese número cuando $n$ se hace grande.

Por ejemplo, si dado un problema a resolver:
- El algoritmo A lo hace en $2n + 1$ pasos,  
- El algoritmo B lo hace en $5n$ pasos,

Ambos **crecen de forma lineal**, porque al aumentar $n$, la influencia de 2, 5 o +1 se vuelve poco importante.

Por eso, en la notación Big O se **simplifican las expresiones**:

- Se ignoran **constantes multiplicativas** (como 2 o 5).
- Se ignoran **constantes aditivas** (como +1), que suelen representan operaciones sueltas (sumas, divisiones, etc).

Entonces, tanto $2n + 1$ como $5n$ se expresan simplemente como **$O(n)$**.

En resumen, lo que realmente importa es **la forma de crecimiento** (lineal, cuadrático, logarítmico …), no el número exacto de operaciones.

⚖️ **¿Y si una parte es más lenta que las otras?**

Supongamos que una de las partes tiene un **ciclo anidado** como este:



```python
i = 0
while i < n:
    j = 0
    while j < n:
        # comparaciones entre precios[i] y precios[j]
        j += 1
    i += 1
```

Esta parte, como ya vimos en el ordenamiento por selección, tiene **complejidad cuadrática**: $O(n^2)$.

Entonces la complejidad total del algoritmo en Big O sería: $O(n^2) + O(n) = O(n^2)$

En la expresión simplificada, **prevalece la parte más lenta** porque es la que más contribuye al tiempo total de ejecución.

📈 **Resumen de tipos de crecimiento en algoritmos**


Hemos visto que cuando analizamos la eficiencia de un algoritmo nos interesa saber cómo crece en tiempo de ejecución (o a veces también el uso de memoria) a medida que aumenta el tamaño de los datos $n$.

En general, las funciones más comunes que aparecen en la notación Big O, ordenadas de menor a mayor crecimiento de tiempo, son:

- $O(1)$ - constante.
- $O(\log n)$ - logarítmico.
- $O(n)$ - lineal.
- $O(n\log n)$ - lineal-logarítmico.
- $O(n^2)$ - cuadrático.
- $O(2^n)$ - exponencial.
- $O(n!)$ - factorial.

Gráficamente, se puede observar cómo crece el tiempo estimado en cada caso a medida que $n$ se incrementa:

<div style="display: flex; justify-content: space-between;"> <img src="https://drive.google.com/uc?id=1FsvWKgT7duy8M3hbk7X1oBwh4ZjsKVyO" width="500" />

En el caso **constante** ($O(1)$) el tiempo de ejecución no depende del tamaño de los datos. El crecimiento **logarítmico** ($O(\log n)$)  es muy lento, y por lo tanto, ideal para algoritmos muy eficientes. En el caso **lineal** ($O(n)$), el tiempo de ejecución crece en proporción directa al tamaño de entrada.

El crecimiento **lineal-logarítmico** ($O(n \log n)$) aparece en muchos algoritmos de ordenamiento eficientes y muy conocidos, como *Quicksort* o *Merge sort*, que no vimos porque requieren técnicas más avanzadas de programación. En el caso **cuadrático** ($O(n^2)$), el tiempo crece mucho más rápido, y suele aparecer cuando hay bucles anidados.

El crecimiento **exponencial** ($O(2^n)$) se vuelve muy costoso incluso con $n$ moderado, ya que duplica el tiempo con cada incremento. Finalmente, el caso **factorial** ($O(n!)$) es el más extremo: tiene un crecimiento rapidísimo, lo que lo hace inviable en problemas grandes.
