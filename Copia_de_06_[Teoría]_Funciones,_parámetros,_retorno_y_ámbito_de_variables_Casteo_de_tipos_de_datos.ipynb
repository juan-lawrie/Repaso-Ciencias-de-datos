Mirá el siguiente código, analizalo línea por línea y luego ejecutalo para ver los resultados:

````python
def calcular_mediana(lista):
    lista_ordenada = lista.copy()
    lista_ordenada.sort()

    n = len(lista_ordenada)
    if n % 2 == 0:
        mediana = (lista_ordenada[n // 2 - 1] + lista_ordenada[n // 2]) / 2
    else:
        mediana = lista_ordenada[n // 2]

    return mediana

ingresos_mensuales = [570, 1200, 590, 500, 600, 1500, 610, 550, 630, 650, 620, 580]

datos = ingresos_mensuales.copy()
datos.sort()
n = len(datos)

# Q2 (mediana)
Q2 = calcular_mediana(datos)

# Separación en mitades
if n % 2 == 0:
    mitad_inferior = datos[:n // 2]
    mitad_superior = datos[n // 2:]
else:
    mitad_inferior = datos[:n // 2]
    mitad_superior = datos[n // 2 + 1:]

# # Medianas de las mitades (Q1 y Q3)
Q1 = calcular_mediana(mitad_inferior)
Q3 = calcular_mediana(mitad_superior)

print("Q1:", Q1)
print("Q2:", Q2)
print("Q3:", Q3)
````

### 🛠️ Funciones propias

En el ejemplo anterior apareció un concepto muy importante en programación: las **funciones**.

Por primera vez, vemos cómo **definir una una función propia** y  cómo **usarla varias veces** sin tener que repetir el mismo código.

En programación, una función es **un bloque de código reutilizable**: se escribe una sola vez y luego se puede (también decimos "llamar") todas las veces que haga falta, con distintos datos.

En el ejemplo, definimos una función que toma una lista, calcula la mediana y devuelve ese valor:

```python
def calcular_mediana(lista):
    lista_ordenada = lista.copy()
    lista_ordenada.sort()
    
    n = len(lista_ordenada)
    
    if n % 2 == 0:
        mediana = (lista_ordenada[n // 2 - 1] + lista_ordenada[n // 2]) / 2
    else:
        mediana = lista_ordenada[n // 2]
    
    return mediana
```

En este código:
- `def` es la **palabra clave** que le dice a la compu que **vamos a definir una función**.
- `calcular_mediana` es el **nombre** que le damos a la función.
- `lista` es un **dato de entrada**, también llamado *argumento*. Si hay más, se separan con comas.
- El **bloque** que sigue está **indentado**, como es habitual en Python.
- `return` indica **el valor que se devuelve** al usar la función.


🎯 **¿Cómo se usa una función?**


Una vez que la escribimos, podemos **usarla tantas veces como necesitemos**.

Eso se hace escribiendo su nombre y pasando los datos entre paréntesis.

Por ejemplo:

```python
Q2 = calcular_mediana(datos)
```

En este caso:
- Se ejecuta el bloque de la función.
- La variable `lista` pasa a valer lo mismo que `datos`.
- El valor resultante se guarda en `Q2`.

Después hicimos dos llamadas más:

```python
Q1 = calcular_mediana(mitad_inferior)
Q3 = calcular_mediana(mitad_superior)
```

En cada llamada, lo único que cambia es **el conjunto de datos que usamos**.

El código que calcula la mediana es siempre el mismo, pero **se aplica a diferentes listas**.


📌 En resumen, una función:
- Tiene un **nombre**.
- Recibe **valores de entrada**.
- Ejecuta una serie de pasos.
- Devuelve un **resultado**.

Se define una sola vez y **se puede usar muchas veces**.

Esto ayuda a escribir **menos código repetido** y a que todo sea **más ordenado y fácil de leer**.

### 🔁 ¿Paso por valor o por referencia?

Cuando usamos funciones en Python, muchas veces **le pasamos datos como argumentos** (por ejemplo, una lista o un número). Pero... ¿qué pasa con esos datos una vez que entran a la función? ¿Se modifican afuera también? 🤔

Para responder esto, necesitamos entender dos ideas clave:

📌 **Paso por valor (valores simples, como números o strings)**

Cuando pasamos **números, textos u otros valores simples** a una función, la compu trabaja con **una copia del valor**. Eso significa que **cualquier cambio dentro de la función no afecta el valor original**.

Por ejemplo:

```python
def sumar_diez(numero):
    numero = numero + 10
    return numero

x = 5
resultado = sumar_diez(x)
print("x:", x)            # sigue siendo 5
print("resultado:", resultado)  # es 15
```

Aunque dentro de la función `numero` cambie, la variable `x` afuera no se modifica. Esto es paso por valor.


📌 **Paso por referencia (listas)**

Cuando pasamos **listas** a una función, la compu **no crea una copia** automáticamente. En su lugar, **trabaja con la referencia al mismo lugar en memoria**, es decir, **apunta a la misma lista**.

Eso significa que **si modificás la lista dentro de la función, también se modifica afuera** 😮

Por ejemplo:

```python
def agregar_valor(lista):
    lista.append(100)

numeros = [1, 2, 3]
agregar_valor(numeros)
print(numeros)  # [1, 2, 3, 100]
```

En este caso, la función **sí modificó la lista original**. Esto es paso por **referencia**.

🎯 **¿Cómo evitamos cambiar una lista sin querer?**

A veces **no queremos modificar la lista original**, sino trabajar con una copia. En ese caso usamos `copy()`:

```python
def ordenar_sin_tocar(lista):
    copia = lista.copy()
    copia.sort()
    return copia

original = [5, 3, 1]
ordenada = ordenar_sin_tocar(original)

print("Original:", original)  # [5, 3, 1]
print("Ordenada:", ordenada)  # [1, 3, 5]
```

Así hicimos en el ejemplo de `calcular_mediana(lista)`:

```python
lista_ordenada = lista.copy()
```

De esta manera, **nos aseguramos de que la lista original quede intacta**, y solo trabajamos con una copia.

Mirá este código y tratá de entenderlo antes de ejecutarlo:

````python
def calcular_cuartiles(lista):
    datos = lista.copy()
    datos.sort()

    n = len(datos)
    if n % 2 == 0:
        mitad_inferior = datos[:n // 2]
        mitad_superior = datos[n // 2:]
    else:
        mitad_inferior = datos[:n // 2]
        mitad_superior = datos[n // 2 + 1:]

    Q1 = calcular_mediana(mitad_inferior)
    Q2 = calcular_mediana(datos)
    Q3 = calcular_mediana(mitad_superior)

    return Q1, Q2, Q3
````

### 🧱 Funciones que usan otras funciones

En programación, es muy común que una función use otra dentro suyo.

En el ejemplo de `calcular_cuartiles`, adentro usamos la función `mediana`.

Esto se puede hacer porque la función mediana **está definida antes**.

```python
def calcular_cuartiles(lista):
    # cuerpo de la función
    ... = calcular_mediana(...)  # llamada a otra función
```

Esta idea nos permite **reutilizar código y organizar mejor nuestras soluciones**.

Cada función es una pieza y podemos combinarlas para resolver tareas más complejas sin repetir el trabajo.

### 🎛️ Parámetros y argumentos

Cuando definimos una función, le damos **valores de entrada que necesita para funcionar**. Estos valores se escriben entre **paréntesis**.

Por ejemplo, en esta definición:

```python
def calcular_cuartiles(lista, redondear=False):
```

- `lista` es un **parámetro obligatorio**: siempre tenemos que darle un valor cuando usamos la función.
- `redondear` es un **parámetro opcional**: tiene un **valor por defecto** (`False`), así que si no escribimos nada, se usará ese valor.

🎯 **¿Qué es un valor por defecto?**

Un valor por defecto es una especie de "valor sugerido". La compu lo usará si no le damos uno al llamar la función.

Por ejemplo:

```python
calcular_cuartiles(ingresos)                    # usa redondear = False
calcular_cuartiles(ingresos, redondear=True)    # usa redondear = True
```

Esto hace que nuestras funciones sean más **flexibles**, ya que podemos adaptalas a distintas situaciones sin tener que escribirlas de nuevo.

🧩 **Tipos de argumentos**

Cuando usamos una función, los valores que le pasamos se llaman **argumentos**.

Hay dos formas principales de pasarlos:
- **Posicional**: se asignan en orden, como en `calcular_cuartiles(ingresos)` → `lista = ingresos`.
- **Nombrado**: indicamos el nombre del parámetro, como en `calcular_cuartiles(ingresos, redondear=True)`.

Podemos combinar ambos tipos, siempre que:
- los **argumentos posicionales vayan primero** y
- **no repitamos nombres**.

Por ejemplo, las siguientes son formas válidas de llamar a una función:

```python
calcular_cuartiles(ingresos)                       # usa el valor por defecto de redondear
calcular_cuartiles(ingresos, redondear=True)      # le damos otro valor a redondear
```

Mientras que éstas llamadas son inválidas y darán error:

```python
calcular_cuartiles(lista=ingresos, True)  # no puede ir un argumento posicional después de uno nombrado
```



### 📦 ¿Qué devuelve una función?

Las funciones pueden **devolver uno más valores**. En el caso de la función `cuartiles`, devuelve tres valores juntos:

```python
return Q1, Q2, Q3
```

Y cuando usamos la función, podemos guardar esos tres resultados así:

```python
q1, q2, q3 = calcular_cuartiles(ingresos)
```

Esto es muy útil cuando queremos calcular **varios resultados al mismo tiempo** con una sola función.

### 🧠 Funciones integradas

Python ya viene con **muchas funciones integradas** (o *built-in functions*, en inglés) que podemos usar directamente, sin tener que definirlas.

En general, son funciones que nos permiten hacer tareas comunes. Desde la primera clase estuvimos usando algunas, como:
- `print(valor)`, para mostrar algo en pantalla.
- `len(lista)`, para calcular la cantidad de elementos (longitud) de una lista.

Existen otras funciones integradas más. En el ejemplo, aparece la función `round()` que, como lo sugiere su nombre, **redondea un número**.

Muchas de estas funciones reciben argumentos, algunos de ellos están definidos con **valores por defecto**.

Este es el caso de la función `round()` que sirve para **redondear un número** a una cantidad de dígitos de precisión.

La documentación oficial de Python de esta función indica:

```python
round(number, ndigits=None)
```

- `number` es un **parámetro obligatorio**: el número que queremos redondear.
- `ndigits` es un **parámetro opcional**: nos permite indicar a cuántos decimales redondear.
Su valor por defecto es `None`.

Así, por ejemplo, para redondear 3.14159 a dos decimales, escribiríamos `round(3.14159, 2)`. Esto daría como resultado 3.14.

📌 `None` es un valor especial en Python que representa “nada” o “ausencia de valor”. Es común verlo como valor por defecto en funciones.

📌 Si `ndigits` se omite, su valor por defecto será `None` y la función retornará el **entero más cercano** al valor de entrada.

🛠️ **Otras funciones integradas útiles**

Además de `round()`, Python ofrece muchas otras funciones que usaremos frecuentemente:
- `sum(lista)`: suma todos los valores de una lista o conjunto.
- `max(lista)` / `min(lista)`: encuentran el mayor o menor valor de un conjunto de datos.
- `sorted(lista)`: devuelve una nueva lista ordenada (no modifica la original).
- `type(valor)`: devuelve el tipo de dato (por ejemplo, si es `int`, `float`, `list`, etc.).
- `list(secuencia)`: convierte la `secuencia` en una lista.
- `range(i,j,k)`: genera una secuencia de enteros de `i` hasta `j-1` con paso `k`.

### 🧹 Errores comunes al usar funciones

Al trabajar con funciones, es importante prestar atención a ciertos detalles para evitar errores.

Cuando creamos una **función propia**, debemos respetar la **sintaxis correcta**:
- Usamos `def` para comenzar la definición.
- El **nombre** de la función va seguido de los **paréntesis** con los **argumentos** (si lo hay).
- Al final de la línea de la función, colocamos **dos puntos** (`:`).
- Todo el **cuerpo de la función** debe estar **indentado**.
- Para devolver un **resultado**, usamos la instrucción `return`.

Cuando usamos **funciones integradas** (como `round()`, `max()`, `len()`, etc.) tenemos que tener cuidado con:
- El **nombre** de la función.
- El **orden y tipo de los argumentos**.

📌 Si tenemos dudas, siempre es bueno revisar la documentación oficial de Python para asegurarnos de usarlas correctamente.

En el caso de `round()`, primero se la pasa el valor a redonder y luego, opcionalmente, el número de dígitos de precisión.

Si cambiamos el orden de los argumentos o pasamos valores incorrectos, podemos obtener errores o resultados inesperados.

🕵️‍♀️ **Usar print para debbuging dentro de la función**

Siempre se puede hacer **debugging** (buscar y corregir errores) agregando instrucciones `print()` dentro de la función.

Esto nos permite ver qué valores están llegando como argumentos o cómo se están calculando los resultados paso a paso.

Por ejemplo:

```python
def suma_doble(a, b):
    print("Valor de a:", a)
    print("Valor de b:", b)
    resultado = (a + b) * 2
    print("Resultado calculado:", resultado)
    return resultado

suma_doble(3, 5)
```

Acá vemos cómo usamos `print()` dentro de la función `suma_doble()` para verificar qué valores se están usando y cuál es el resultado antes de devolverlo.

Así podemos entender mejor el comportamiento del código si algo no está funcionando como esperábamos.

📌 Una vez que nos aseguramos de que todo funciona correctamente, podemos quitar los `print()` para dejar la función más limpia.

### 🎯 Ámbito de las variables: ¿dónde existen?

Cuando definimos una variable dentro de una función, esa variable **solo existe ahí adentro**.

Se dice que tiene **ámbito local**. No podés acceder a ella desde afuera de la función.

Por ejemplo:

```python
def calcular_desvio(lista):
    media = calcular_media(lista)
    diferencias = []
    for ingreso in lista:
        diferencias.append((ingreso - media) ** 2)
    desvio = (sum(diferencias) / len(diferencias)) ** 0.5
    return desvio

desvio_ingresos = calcular_desvio(ingresos)
print(diferencias)  # da error porque 'diferencias' no está definida afuera de la función
```

El error se da porque `diferencias` fue definida **dentro de la función**, así que existe solo durante la ejecución de esa función.

Fuera de la función, la compu no sabe qué es esa variable porque no existe.

🔍 **Variables locales vs. globales**

Una **variable local** se crea **dentro de una función** y **solo existe ahí**. No podés usarla afuera.

Por otro lado, una **variable global** se define **fuera de cualquier función**, y podés usarla en todo el programa (incluso dentro de funciones).

Por ejemplo:

```python
tasa = 0.21  # global

def calcular_precio(precio_base):
    impuesto = precio_base * tasa  # impuesto es local
    total = precio_base + impuesto
    return total

precio = 1000  # global
print(calcular_precio(precio))
print(impuesto)  # da error: 'impuesto' es local a la función
```

Para seguir **buenas prácticas**:
- Evitá usar el mismo nombre para una **variable local** y una **global**. Aunque técnicamente es posible, puede generar confusiones.
- Si necesitás usar una variable global dentro de una función podés hacerlo, pero no modificarlas. En general, **es mejor pasar los valores como argumentos** y devolver los resultados.




📌  Pistas para saber si una variable es local o global
- Si está **definida dentro de la función**, es **local**.
- Si está **definida fuera de la función**, es **global**.
- Los **parámetros de la función** también son locales: existen solo dentro de la función.

⚠️ **Evitá pisar funciones con nombres de variables**

Un error muy común en programación es usar **el mismo nombre para una variable que para una función**.

Recordemos que **las funciones también tienen nombres**, igual que las variables. Ese nombre es su *identificador*, y como tal, **es único**.

Si más adelante usás ese mismo nombre para crear una variable, vas a estar **pisando la función**: la reemplazás con la variable, y ya no podés usar la función como antes.

Por ejemplo, supongamos que definimos la función `calcular_media()` y luego hacemos esto:

```python
calcular_media = [1, 2, 3]   # esto pisa la función calcular_media()
```

Después de eso, si intentás llamar a la función `calcular_media()`:

```python
calcular_media([400, 600, 800])  # da error porque calcular_media ya no es una función
```

No va a funcionar porque el nombre `calcular_media` ya no apunta a la función, sino a la lista que creaste.

📌 Por eso, **es una buena práctica evitar usar como nombres de variables**:
- el nombre de **funciones que definiste**, como `calcular_media`, `calcular_desvio`, etc.
- nombres de **funciones integradas** de Python, como `sum`, `max`, etc.

Usar nombres claros y distintos para cada cosa ayuda a evitar errores difíciles de detectar y hace tu código más fácil de leer.

### 📥 Funciones sin valor de retorno

En programación, una **función sin valor de retorno** es aquella que **realiza una acción** pero **no devuelve ningún valor** al lugar donde se llamó.

En otras palabras, la función **no tiene una instrucción `return`** que envíe un valor de vuelta.

Por eso, **no se puede guardar su resultado en una variable**, ni usar ese valor más adelante.

Cuando definimos una función que no retorna un valor, la sintaxis básica es la siguiente:

```python
def nombre_funcion():
    # Cuerpo de la función: realiza operaciones o imprime resultados
    print("Esta es una función sin retorno")
```

En el ejemplo, la función `resumen()` calcula algunos valores (media y desvío), pero no los devuelve. Solo imprime el resultado en pantalla:

```python
def resumen(lista):
    media = calcular_media(lista)
    desvio = calcular_desvio(lista)

    print("Media:", media)
    print("Desvío:", desvio)

resumen(ingresos)
```

Si intentamos guardar el resultado de esta función:

```python
resultado = resumen(ingresos)
print(resultado)
```

La salida será:

```python
Media: ...
Desvío: ...
None
```

Esto ocurre porque, al no haber un `return`, la compu devuelve automáticamente el valor especial `None`, que indica **ausencia de retorno**.

Otro ejemplo de este tipo de funciones lo vimos al comienzo cuando analizamos el paso por referencia y la modificación de listas:

```python
# Función sin valor de retorno
def agregar_valor(lista):
    lista.append(100)

numeros = [1, 2, 3]
agregar_valor(numeros)  # LLamada sin asignar el resultado a una variable
print(numeros)  # [1, 2, 3, 100]
```

En este caso, la función `agregar_valor` no tiene valor de retorno, por eso se llama sin asignar el resultado a una variable.

📌 En resumen:
- **Funciones con retorno**: devuelven un valor con `return`, que se puede guardar o usar después.
- **Funciones sin retorno**: no devuelven valor. Solo realizan acciones como imprimir o mostrar información.

🔄 **`print` no es lo mismo que `return`**

Cuando usás `print()` dentro de una función, **solo mostrás algo en pantalla**.
Ese valor **no queda guardado ni se puede usar después**.

Cuando usás `return`, **la función entrega un resultado**, y ese valor **podés guardarlo** en una variable o usarlo en otro cálculo.

Para ejemplificar esta situación vamos a tomar un subconjunto de los datos con los que veníamos trabajando.

Fijate este código y analizalo antes de ejecutarlo:

````python
nombres = ["Ana", "Bruno", "Clara", "Diego", "Esteban"]
ingresos_como_texto = ["570", "1200", "590", "500", "600"]

ingresos_como_numero = []
for ingreso_str in ingresos_como_texto:
    ingreso_num = int(ingreso_str)
    ingresos_como_numero.append(ingreso_num)

cantidad_mayores_a_1000 = 0
for i in range(len(nombres)):
    nombre = nombres[i]
    ingreso = ingresos_como_numero[i]

    gana_mas_de_1000 = ingreso > 1000
    cantidad_mayores_a_1000 += gana_mas_de_1000

    mensaje = nombre + " gana más de $1000?: " + str(gana_mas_de_1000)
    print(mensaje)

print("Cantidad de personas que ganan más de $1000:", cantidad_mayores_a_1000)
````

### 🧪 Casteo de tipos de datos

Cuando trabajamos con datos a veces necesitamos **convertir un valor de un tipo a otro**.

Esto se llama **casteo** (o *type casting*, en inglés). Existen dos formas de casteo: explícito e implícito.

🔁 **Casteo explícito**

Es cuando **le pedimos a la compu que convierta un dato de un tipo a otro**, usando funciones como:
- `int(valor)` → convierte a entero
- `float(valor)` → convierte a decimal
- `str(valor)` → convierte a texto
- `bool(valor)` → convierte a booleano

Por ejemplo, en el código anterior lo vemos en estos casos:

```python
ingreso_num = int(ingreso_str)  # casteo explícito de str a int
mensaje = nombre + " gana más de $1000?: " + str(gana_mas_de_1000)  # casteo explícito de bool a str
```

Acá hay dos situaciones:
- **Convertir de texto a número**: `ingreso_str` contiene un valor de tipo cadena. Al usar la función `int()`, estamos convirtiendo ese valor en un número entero. Por ejemplo, `"570"` se convierte en `570`.
- **Convertir de booleano a texto**: `gana_mas_de_1000` es un valor booleano (`True` o `False`). Para incluirlo en una cadena de texto, lo convertimos a texto usando `str()`. Por ejemplo, `True` se convierte en `"True"`.


🔄 **Casteo implícito**

Es cuando **la compu convierte los datos automáticamente**, sin que lo digamos, porque **es seguro y tiene sentido hacerlo**.

Esto ocurre en operaciones matemáticas, donde los tipos son compatibles y la conversión no genera ambigüedades.

En el ejemplo anterior lo vemos en este caso:

```python
cantidad_mayores_a_1000 += gana_mas_de_1000
```

Acá, `gana_mas_de_1000` es un valor `bool` (como `True`), y cuando lo sumamos con un número, **la compu lo convierte implícitamente**:
- `True` se convierte en `1`
- `False` se convierte en `0`

Entonces, aunque no escribimos `int(gana_mas_de_1000)`, la compu hace por nosotros.

Otro caso de casteo implícito se da cuando sumamos enteros y flotantes:

```python
entero = 3
decimal = 2.5
resultado = entero + decimal  # la compu convierte 3 a 3.0 automáticamente
print(resultado)  # muestra 5.5
```
La compu convierte el entero 3 en el flotante 3.0 para poder hacer la suma correctamente.

El resultado de mezclar un entero con un flotante es un `float`.

🧠 **¿Por qué importa el casteo?**

- **Evita errores**: si no convertimos correctamente los tipos de datos, el programa puede fallar (por ejemplo, no se puede sumar un número con texto).
- **Trabaja con datos reales**: los datos que recibimos pueden estár en formato de texto, y entender el casteo nos permite manejarlos adecuadamente.
- **Flexibilidad**: la compu convierte automáticamente tipos compatibles, pero es importante saber qué tipo de dato estás manejando para evitar resultados inesperados.
