1. Recorre la lista y busca el valor mínimo desde la posición actual hasta el final.
2. Intercambia el valor mínimo con el valor de la posición actual.
3. Repite el proceso para el resto de la lista.

Cuando lo aprendimos, lo implementamos usando ciclos `while`:

```python
lista = [64, 25, 12, 22, 11]
n = len(lista)
i = 0

while i < n - 1:
    min_idx = i
    j = i + 1

    while j < n:
        if lista[j] < lista[min_idx]:
            min_idx = j
        j += 1

    lista[i], lista[min_idx] = lista[min_idx], lista[i]
    i += 1

print("Lista ordenada:", lista)
```
