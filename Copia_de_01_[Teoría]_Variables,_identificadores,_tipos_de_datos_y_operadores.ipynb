### 🏷️ Identificadores

Cada variable tiene un nombre que la distingue de las demás en el programa. A este nombre lo llamamos **identificador**, y debe ser único dentro del programa.

Sin embargo, no todos los caracteres o símbolos pueden usarse en los identificadores. Algunos símbolos tienen un propósito especial en el lenguaje de programación, por lo que no pueden ser parte del nombre de una variable. Por ejemplo, el signo menos (`-`), como veremos más adelante, se usa para la resta, por lo que no podemos usarlo en un identificador.

Las reglas básicas para los identificadores son:
- ✅ Pueden contener **letras** (mayúsculas y minúsculas).
- ✅ Pueden contener **números**, siempre que **no estén al comienzo**.
- ✅ Pueden contener algunos pocos **símbolos**, como el guión bajo (`_`), que usamos comúnmente para separar palabras (por ejemplo, `es_mayor`).

Algo para tener en cuenta es que en la mayoría de los lenguajes de programación, `edad` y `Edad` son identificadores **diferentes**. Es decir, para la compu las minúsculas son símbolos diferentes de las mayúsculas, así que si hacemos `edad = 25` y luego `Edad = 30`, estos dos números que asignamos se guardarán en **dos posiciones distintas** de la memoria.


Finalmente, vale aclarar que hay algunas palabras que no pueden usarse como identificadores porque
tienen un significado especial para el lenguaje de la compu. Estas palabras, llamadas **palabras reservadas**, incluyen `print`, y otras que ya iremos aprendiendo, como `if`, `else`,`for`, `in`, `not`, etc.

### 🔠 Tipos de datos

En programación, los datos que manejamos tienen diferentes **tipos**, lo cual determina las operaciones que podemos realizar con ellos.

Los **tipos de datos** más comunes son:

- **Números enteros (int):** como `25`, `0`, `-5`.
- **Números decimales o flotantes (float):** como `1.71`, `-2.5`, `0.001`.
- **Cadenas de texto o strings (str):** como `"Ana"`, `"Santa Fe"`,  `"CP 3000"`.
- **Booleanos (bool):** sólo pueden tomar dos valores, `True` (verdadero) o `False` (falso).

Los enteros representan números sin decimales, mientras que los flotantes representan números con decimales, ambos se usan para realizar operaciones matemáticas. Algo a tener en cuenta es que en la compu los números flotantes se escriben con **punto**, no con coma. Por lo tanto, para representar `3,14` debemos escribir `3.14`.

Las cadenas de texto pueden contener letras, números y otros caracteres. Se escriben entre **comillas** simples o dobles, siempre usando el mismo tipo de comilla para abrir y cerrar la cadena. Por ejemplo, `"Ana"` y `'Ana'` son formas equivalentes de definir la misma cadena.

Aunque una cadena puede contener números, como `"2025"`, la compu siempre las interpreta como **texto**, por lo que no se pueden usar directamentente en operaciones matemáticas.

Los booleanos son útiles para representar condiciones de verdadero o falso y para guardar datos que son de tipo **Sí/No**. Por ejemplo, `es_mayor = True` indica que una persona es mayor de edad, mientras que `usuario_registrado = False` señala que un usuario no está registrado en un sistema. También se pueden obtener a partir de comparaciones entre números o cadenas, como veremos más adelante.

Para conocer el tipo de dato asociado a una variable podemos usar la función `type()`:

````python
print(type(nombre))
print(type(edad))
print(type(altura))
print(type(es_mayor))
````

### 🔣 Operadores

Con las variables logramos que la compu haga cuentas. Hicimos sumas pero también hicimos otras operaciones.

Por ejemplo, cuando escribimos `edad_actual = 20` estamos usando el **operador de asignación** `=` para asignar el número `20` a la variable `edad_actual`. Cuando se hace una asignación, el valor queda guardado en una posición de la memoria.

Los operadores también permiten usar lo que guardamos en las variables para obtener otros resultados. En el ejemplo anterior, usamos el operador aritmético `+` para sumarle 5 al valor guardado en la variable `edad_actual`, y el resultado de esa operación también lo guardamos en una variable `edad_futura`, haciendo otra asignación con el operador `=`.

Como te imaginarás, existen muchos otros operadores para hacer cuentas matemáticas! Se llaman **operadores aritméticos** y los más comunes son:

|Operador|Descripción|Ejemplo|Resultado|
|:--|:--|:--|:--|
|`+`|Suma|`7 + 2`|`9`|
|`-`|Resta|`7 - 2`|`5`|
|`*`|Multiplicación|`7 * 2`|`14`|
|`/`|División|`7 / 2`|`3.5`|
|`//`|División entera|`7 // 2`|`3`|
|`%`|Módulo (resto)|`7 % 2`|`1`|
|`**`|Potencia|`7 ** 2`|`49`|

En resumen, usamos `+` para sumar, `-` para restar, `*` para multiplicar, `/` para dividir con decimales, `//` para dividir obteniendo solo la parte entera, `%` para obtener el resto de una división y `**` para calcular potencias.

Más adelante veremos otros operadores, como los **relacionales** y los **lógicos** que nos ayudarán a escribir programas más avanzados. 🚀


Como vemos en la tabla, tenemos tres tipos de división: `/`, `//` y `%`.

Si recordás, en una división intervienen cuatro componentes principales:
- *Dividendo*: el número a dividir.
- *Divisor*: el número por el cual se divide.
- *Cociente*: el resultado de la división.
- *Resto o residuo*: lo que sobra después de dividir.

Para obtener el cociente podemos usar `/ `o `//` dependiendo del caso, y para el resto `%`.

Para verlo, tomemos como ejemplo dividir 7 entre 2. Mirá este código y luego ejecutálo para obtener el resultado:

````python
dividendo = 7
divisor = 2

cociente = dividendo / divisor   # 3.5 (división normal)
cociente_entero = dividendo // divisor  # 3 (división entera)
resto = dividendo % divisor   # 1 (resto de la división)

print(cociente)
print(cociente_entero)
print(resto)
````

El operador `/` (división normal) devuelve el **resultado completo de la división**, incluyendo la parte decimal.

```python
7 / 2  # 3.5
```

En este caso, la parte entera es 3 y la parte decimal es 0.5.

Este operador siempre devuelve un `float`, incluso si el resultado es exacto (por ejemplo, `6 / 2` da 3.0).

Por otro lado, el operador `//` (división entera) devuelve **solo la parte entera del cociente**, descartando la parte decimal.

```python
7 // 2  # 3
```

Esto responde a la pregunta: ¿cuántas veces completas entra el 2 en el 7? La respuesta es: 3 veces.

Acá el tipo del resultado depende de los operandos: si ambos son `int`, el resultado es `int`; si alguno es `float`, el resultado es `float`.

📌 **Importante: la división entera no es lo mismo que redondear**.

Finalmente, el operador `%` (módulo) devuelve el **resto de la división entera**.

```python
7 % 2  # 1
```
Esto se debe a que 7 = 2 x 3 + 1, donde 1 es el resto.

Este operador también sigue la misma regla de tipos que `//`.

Para cualquier operación aritmética (suma, multiplicación, etc.) la regla general para el tipo de dato del resultado será:
- Si **ambos** valores son `int`, el resultado es `int`, **excepto en `/` y en exponentes negativos**.
- Si **al menos** un valor es `float`, el resultado es `float`.

🔤 **Operadores `+` y `*` con cadenas**

Además de los números, vimos que algunos operadores también se pueden usar con cadenas de texto.

Por ejemplo, el operador `+` se usa para **concatenar** (unir) dos o más cadenas. Cuando hacemos `nombre + " " + apellido` estamos concatenando las cadenas "`Juan"`, `" "` (un espacio) y `"Pérez"` para formar la cadena "Juan Pérez", que guardamos en la variable `nombre_completo`.

Como habrás notado, el espacio en blanco (que normalmente hacemos con la barra espaciadora del teclado) es un caracter más, y en el ejemplo se incluye explícitamente para separar el nombre y el apellido. Si omitiéramos el espacio y escribiéramos sólo `nombre + apellido`, el resultado sería `"JuanPérez"`, sin el espacio entre el nombre y el apellido.

Cabe remarcar la diferencia entre la cadena vacía `""` y la cadena " " que contiene un caracter, que es el espacio.

El otro operador que no vimos en el ejemplo pero que es útil con cadenas es el `*`. Aunque en los números este operador representa la multiplicación, en el caso de las cadenas permite **repetir** un texto. Por ejemplo, `"Juan" * 3` da como resultado `"JuanJuanJuan"`, repitiendo la cadena `"Juan"` tres veces.

### 📝 Comentarios

En el ejemplo vemos también algunas líneas en color verde que comienzan con el símbolo `#`.

En programación, estas líneas se llaman **comentarios** y se utilizan para agregar notas que explican una línea de código o partes de un programa. También pueden usarse para desactivar temporalmente partes del código sin necesidad de eliminarlas.

Es una buena práctica incluir comentarios para aclarar partes complejas de un programa, pero hay que evitar caer en redundancias. Antes de agregar un comentario, vale la pena preguntarnos si el código sería más comprensible mejorando los nombres de las variables o dividiendo un cálculo complejo en pasos más simples.

### ➡️ Indentación

Parece que a la compu no le gusta que las instrucciones que le damos estén desacomodadas 🤔.

Cuando asignamos

```python
    apellido = "Gómez"
```

le pusimos adelante dos espacios más que en la línea anterior, que era

```python
nombre = "María"
```

y estaba bien alineada contra el margen izquierdo, ¿no?

En programación, estos espacios al inicio de una línea se conocen como **indentación**.

La indentación es una característica fundamental en Python 🐍, ya que le permite a la compu entender la estructura del código y determinar qué instrucciones pertenecen a qué parte del programa.

En este caso, no era necesario indentar la asignación de `apellido`, ya que ambas líneas son independientes y no forman parte de una estructura que requiera indentación. Sin embargo, más adelante veremos ejemplos donde la indentación es obligatoria para que el código funcione correctamente.

A diferencia de Python, en otros lenguajes de programación, la indentación es opcional. Esto significa que, aunque el código esté mal indentado, el programa puede funcionar igual. Sin embargo, es una buena práctica indentar correctamente, ya que mejora la legibilidad del código y reduce los errores. Python, al exigir la indentación, nos obliga a escribir código más claro y organizado desde el principio, lo que es una gran ventaja para desarrollar buenos hábitos de programación.

### 🐞 Errores detectados y bugs

Cuando programamos, es común cometer errores. Algunos son fáciles de identificar porque la compu los detecta y nos muestra un **mensaje de error**.

Por ejemplo, los errores de indentación que vimos antes son detectados automáticamente. Otro error común ocurre cuando olvidamos cerrar las comillas en una cadena de texto. En este caso, la compu no sabe dónde termina el texto y nos muestra un mensaje de error en pantalla.

También pueden aparecer errores al usar identificadores no válidos, como en `edad actual` o en `2edad_futura`. La compu sabe que los identificadores no pueden tener espacios ni comenzar con un número y nos lo hace saber.

Otro ejemplo de error se da cuando cuando escribimos mal un identificador, como `apelido` en lugar de `apellidomensaje`, o `Nombre_completo` en lugar de `nombre_completo`. La compu nos avisa que no puede encontrar una variable definida con ese nombre.

Por último, al usar funciones, es importante recordar cerrar los paréntesis después de pasarle los datos que necesitan. Si olvidamos hacerlo, como en `print(edad_futura` también veremos un mensaje de error.

Muchos de estos errores son ejemplos de errores de **sintaxis**, lo que significa que no seguimos las reglas del lenguaje de programación. La sintaxis define, entre otras cosas, cómo nombrar variables, cómo escribir expresiones y cómo organizar el código de manera que la compu pueda entenderlo correctamente.

🐞 **Los "bugs"**

También hay errores más difíciles de encontrar. Son errores que **no dan error** cuando ejecutamos el código, porque la compu no tiene forma de darse cuenta que algo está mal.

Muchas veces a este tipo de errores se les llama **"bug"** (que significa "bicho"), porque cuenta la historia que en una de las primeras computadoras había un bicho atrapado en los circuitos que hacía que ande todo mal.

En el ejemplo, cuando escribimos un `-` en lugar de un `+` indudablemente la compu no sabe que queríamos sumar, por lo que no tiene forma de indicarnos que ese `-` está equivocado. El "bug" en realidad lo puso quién haya cometido ese error de signo al escribir el programa.

Para evitar los bugs, es importante revisar nuestro código, probar con diferentes valores y usar `print()` para verificar los resultados intermedios. Este proceso de identificar y corregir errores se llama **depuración** del código, o en inglés, *debugging*.

### 🎯 La media

Cuando sumamos las alturas de un grupo, obtenemos un valor total, como 6.76 m. Sin embargo, este número no refleja la altura de ninguna persona en particular. Aunque la suma nos muestra la altura total del grupo, no nos da una idea precisa sobre la altura de cada persona.

Por eso, para obtener un valor más representativo, dividimos la suma de las alturas por el número de personas del grupo. El resultado de esta operación es la **media aritmética** (o simplemente la **media**), un valor que, en este caso, es 1.69 m. Este valor nos da una idea general de la "altura promedio" de las personas en el grupo.

$$
\begin{array}{c c c c c c c}
    & \text{Juana} & \text{Pedro} & \text{Media} & \text{Gisela} & \text{Jorge} &  \\
    %\hline
     & \bullet & \bullet & \color{red} \bullet & \bullet & \bullet & \\
    & 1.65 & 1.68 & \mathbf{1.69} & 1.71 & 1.72 &
\end{array}
$$

La media es una **medida resumen** (es decir, un **número**) que nos proporciona información acerca del **promedio** o **valor central** de un conjunto de datos. Representa la posición alrededor de la cual tienden a agruparse los datos.

Al dividir la suma por el número de personas, la media refleja el tamaño del grupo. Si añadimos más personas al grupo, la media cambiará de acuerdo con la cantidad de personas nuevas, ya que estamos dividiendo por un número mayor. Esto asegura que la media siempre sea un valor que tiene sentido para representar las alturas del grupo.

Más adelante, veremos otros métodos para calcular el promedio o centro de un conjunto de datos. También aprenderemos a describir la dispersión de un conjunto de datos, es decir cuán lejos tienden a estar los datos de su centro.🚀

### 📝 Expresiones

En todos estos ejemplos hemos estado haciendo cálculos mediante **expresiones**.

Una expresión está formada por **operandos** relacionados por **operadores**.

Cuando estas expresiones se evalúan generan un **resultado**. Por ejemplo:

```python
(altura_juana + altura_jorge + altura_gisela + altura_pedro) / 4
```

Un operando puede ser un número, una cadena u otro tipo de dato, y puede estar dado por una **constante**, como el `4` en el ejemplo anterior, o dentro de una **variable** como `altura_juana`.

Ejecutando la siguiente celda se puede obtener el resultado de una expresión muy simple:



````python
1 + 2
````

En este caso, usamos el operador suma, es decir el `+`, para relacionar dos operandos que son las constantes `1` y `2`.

Las constantes también se llaman **literales** porque son valores que se escriben directamente en el código y no provienen de una variable o del resultado de una expresión.



Como ya vimos, un operador importante es el de **operador de asignación**`=`, que ya hemos estado usando mucho en los ejemplos. Simplemente cuando queremos *guardar* el resultado de una expresión en una variable, usamos el operador de asignación.

Es decir, cuando se hace una asignación, el resultado de la expresión es almacenado en alguna posición específica de la memoria de la computadora, para que podamos usarlo de nuevo en nuestro programa, sin tener que volver a calcularlo.


En la siguiente celda *guardamos* la altura de Juana en la variable `altura_juana`:

````python
altura_juana = 1.65
````

Cuando **asignamos múltiples variables**, podemos hacerlo en varias líneas, como en el código donde definimos las alturas de las cuatro personas, o en una sola línea separando cada asignación con una coma.

Ejecutá las dos celdas que siguen para ver que ambas formas dan el mismo resultado:

````python
# Asignación en líneas separadas
altura_juana = 1.65
altura_jorge = 1.72

print(altura_juana)
print(altura_jorge)
````

````python
# Asignación en una sola línea
altura_juana, altura_jorge = 1.65, 1.72

print(altura_juana)
print(altura_jorge)
````

Otra cuestión importante es el **intercambio** de valores entre variables. La forma clásica requiere el uso de una **variable temporal** para evitar perder los valores originales. Sin embargo, también podemos realizar este intercambio utilizando la asignación múltiple que vimos arriba.

Ejecutando las dos celdas de abajo podés probar ambos métodos:

````python
altura_juana = 1.65
altura_jorge = 1.72

# Intercambio con variable temporal
temp = altura_juana
altura_juana = altura_jorge
altura_jorge = temp

print(altura_juana)
print(altura_jorge)
````

````python
altura_juana = 1.65
altura_jorge = 1.72

# Intercambio en una sola línea
altura_juana, altura_jorge = altura_jorge, altura_juana

print(altura_juana)
print(altura_jorge)
````

Por otro lado, a veces necesitamos **actualizar** el valor de una variable aplicándole una operación con una constante. Esto implica tomar el valor actual de la variable, realizar la operación y luego **reasignar** el resultado a la misma variable.

Para simplificar este proceso, Python nos ofrece los **operadores de asignación compuesta**, como `+=`, `-=`, `*=`, y `/=`.

Ejecutá las siguientes celdas para comprobar que ambas formas son equivalentes:

````python
altura_juana = 1.65

# Realizar la operación y reasignar
altura_juana = altura_juana + 0.02

print(altura_juana)
````

````python
altura_juana = 1.65

# Usando el operador de asignación compuesta
altura_juana += 0.02

print(altura_juana)
````

🔢 **Precedencia de operadores aritméticos**

Al igual que cuando hacemos cuentas en matemática, la computadora resuelve las expresiones aplicando los operadores en un determinado orden de prioridad, lo que se conoce como **precedencia** de operadores.

La regla básica es:
1. *Paréntesis* `()`
2. *Potenciación* `**`
3. *Multiplicación y división* `*/`
4. *Suma y resta* `+-`

Siempre que haya **paréntesis**, se resolverá primero lo que está entre paréntesis, sin importar el operador que haya fuera.  
De hecho así lo hicimos en los ejemplos anteriores para indicar que primero queríamos que se hagan las sumas, y luego la división.

Si hay operadores con la misma prioridad en una expresión se resuelven de **izquierda a derecha**, en el orden en el que aparecen en la expresión. Por ejemplo, la suma y la resta tienen la misma prioridad, al igual que la multiplicación y la división. Así que, en la expresión `4 + 5 - 2` primero se realiza la suma `4 + 5` y luego la resta `9 - 2`. En la expresión `8 / 2 * 3` primero se realiza la división `8 / 2` y luego la multiplicación `4 * 3`.
