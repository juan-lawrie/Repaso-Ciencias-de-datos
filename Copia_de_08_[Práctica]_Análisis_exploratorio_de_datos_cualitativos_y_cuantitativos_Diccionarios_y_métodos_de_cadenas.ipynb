Si observás los datos en la lista, vas a ver que hay inconsistencias en mayúsculas/minúsculas y espacios adicionales.

Para analizarlos correctamente, primero debemos limpiarlos.

Completá el código para **eliminar los espacios en los extremos y pasar las respuestas a mayúsculas**:

````python
ocupacion_limpio = []

for r in ocupacion:
    r_limpio = r.strip()
    r_limpio = r_limpio.upper()
    ocupacion_limpio.append(r_limpio)

print("Datos limpios:", ocupacion_limpio)
````

Ahora nos enfocaremos en el resumen numérico.

Calculá las frecuencias absolutas y relativas de cada categoría de ocupación:

````python
frecuencias_absolutas = {}

for r in ocupacion_limpio:
    if r not in frecuencias_absolutas:
        frecuencias_absolutas[r] = 1
    else:
        frecuencias_absolutas[r] += 1

# Mostrá las frecuencias absolutas
print("Frecuencias absolutas:")
for k in frecuencias_absolutas:
    print(k, frecuencias_absolutas)
````

````python
#n = sum(frecuencias_absolutas.values())
n = len(ocupacion_limpio)
frecuencias_relativas = frecuencias_absolutas.copy()

for k in frecuencias_absolutas:
    frecuencias_relativas[k] = frecuencias_absolutas[k]/n

# Mostrá las frecuencias relativas redondeadas a dos decimales
print("Frecuencias relativas:")
for k in frecuencias_relativas:
    print(k, round(frecuencias_relativas[k], 4))
````

Los mismos corresponden a la cantidad total de casos de Dengue y Zika registrados desde la última semana epidemiológica de 2020 hasta la semana 25 de 2021 (es decir, un total de 26 semanas epidemiológicas):

````python
total_casos = [
    4, 10, 17, 35, 50, 95, 102, 103, 129, 110, 217,
    214, 156, 198, 240, 308, 394, 459, 361, 242,
    112, 125, 91, 50, 24, 4
]
````

Usá las funciones definidas debajo para calcular los 5 números resumen:
- Mínimo
- Primer cuartil (Q1)
- Segundo cuartil (Q2)
- Tercer cuartil (Q3)
- Máximo


````python
def calcular_mediana(lista):
    lista_ordenada = lista.copy()
    lista_ordenada.sort()

    n = len(lista_ordenada)

    if n % 2 == 0:
        mediana = (lista_ordenada[n // 2 - 1] + lista_ordenada[n // 2]) / 2
    else:
        mediana = lista_ordenada[n // 2]

    return mediana

def calcular_cuartiles(lista):
    datos = lista.copy()
    datos.sort()

    n = len(datos)
    if n % 2 == 0:
        mitad_inferior = datos[:n // 2]
        mitad_superior = datos[n // 2:]
    else:
        mitad_inferior = datos[:n // 2]
        mitad_superior = datos[n // 2 + 1:]

    Q1 = calcular_mediana(mitad_inferior)
    Q2 = calcular_mediana(datos)
    Q3 = calcular_mediana(mitad_superior)

    return Q1, Q2, Q3
````

````python
minimo = min(total_casos)
maximo = max(total_casos)
q1, q2, q3 = calcular_cuartiles(total_casos)

print("Mínimo:", minimo)
print("Q1:", q1)
print("Q2:", q2)
print("Q3:", q3)
print("Máximo:", maximo)
````

````python
def calcular_media(lista):
    return sum(lista) / len(lista)

def calcular_desvio(lista):
    media = calcular_media(lista)

    suma_diferencias = 0
    for x in lista:
        dif = x - media
        suma_diferencias += dif ** 2

    desvio = (suma_diferencias / len(lista)) ** (1/2)
    return desvio
````
