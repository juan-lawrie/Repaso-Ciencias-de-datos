## 🔍 Iterando listas de otra forma

````python
ingresos_mensuales = [570, 1200, 590, 500, 600, 1500, 610, 550, 630, 650, 620, 580]

for ingreso in ingresos_mensuales:
    print(ingreso)
````

### 🔄 Ciclo for

El ciclo `for` es una estructura de control que nos permite **repetir una acción** para cada elemento de una secuencia, como una lista.

En particular, es útil cuando **ya sabemos cuántas veces queremos repetir algo**, por ejemplo, recorrer una lista completa.

Por ejemplo:

```python
for ingreso in ingresos_mensuales:
    print(ingresos)
```

Este código hace lo siguiente:
- Recorre uno a uno los elementos de la lista `ingresos_mensuales`.
- En cada vuelta del ciclo, la variable `ingreso` toma el valor del siguiente elemento.
- Ejecuta la acción indicada (mostrarlo por pantalla).

Este código se puede leer como: *Para cada ingreso en la lista de ingresos, hacé lo siguiente...*.

📌 Algunos detalles importantes son:
- No es necesario preocuparse por los índices (las posiciones). La compu se encarga de eso.
- Es más claro y directo que `while` para recorrer listas.
- El nombre de la variable (`ingreso`) puede ser cualquiera, pero conviene que tenga sentido según lo que estás recorriendo.

Por otro lado, si queremos recorrer una lista por **posiciones** (índices), podemos combinar `for` con `range()`:

```python
for i in range(0, len(ingresos_mensuales)):
    print("Ingreso en posición", i, ":", ingresos_mensuales[i])
```

Acá, `range(len(ingresos_mensuales))` crea una **secuencia de números** desde 0 hasta `len(ingresos_mensuales) - 1`.

Cuando se quiere una secuencia que comience por el 0, este se puede omitir y se obtiene el mismo resultado:

```python
for i in range(len(ingresos_mensuales)):
    print("Ingreso en posición", i, ":", ingresos_mensuales[i])
```

### ✅ ¿Cómo se escribe un for?

Todo ciclo `for` se escribe con esta estructura:

```python
for variable in secuencia:
    instrucción_1
    instrucción_2
```

Al igual que con el `while`, tiene dos reglas importantes:
- Lleva **dos puntos** (`:`) al final de la línea del `for`.
- Las intrucciones dentro del ciclo debe estar **indentadas** (es decir, con espacios al inicio):

En el código del ejemplo:

```python
for i in range(1, len(ingresos_mensuales))
print("Ingreso en posición", i, ":", ingresos_mensuales[i])
```

Faltan los dos puntos al final del `for` y la línea dentro del `for` no está indentada. Lo correcto sería así:

```python
for i in range(1, len(ingresos_mensuales)):
    print("Ingreso en posición", i, ":", ingresos_mensuales[i])
```

Además, tenemos un *bug* porque la secuencia de índices generada por `range()` comienza en 1, cuando debería comenzar en 0. Por lo tanto, deberíamos escribimos `range(0, len(ingresos_mensuales))` o simplemente `range(len(ingresos_mensuales))` para poder recorrer la lista de principio a fin.

### 📏 ¿Qué hace range?

La función `range()` genera una **secuencia de números enteros**.

Las formas más comunes de usar esta función son:
- `range(b)` → va de `0` hasta `b-1`.
- `range(a, b)` → va de `a` hasta `b-1`.
- `range(a, b, paso)` → como antes, pero de a `paso`.

📌 Como vemos, el valor el extremo derecho de la secuencia, `b`, nunca se incluye, sino que va hasta `b-1`.

Por ejemplo:

```python
for i in range(3, 7):
    print(i)
```

Esto imprime `3 4 5 6`.


Si usamos un solo número, como `range(7)`, empieza desde 0:

```python
for i in range(7):
    print(i)
```

Esto imprime `0 1 2 3 4 5 6`.

También podemos generar una secuencia con un **paso** determinado, por ejemplo de dos en dos:

```python
for i in range(3, 7, 2):
    print(i)
```

Esto imprime `3 5`.

🧰 **¿Para qué combinamos range() con len()?**

Cuando queremos recorrer una lista usando sus **posiciones** (índices), usamos:

```python
for i in range(len(lista)):
    ...
```

Esto es útil porque `len(lista)` calcula cuántos elementos tiene la lista y `range(len(lista))` genera la secuencia de índices desde `0` hasta `len(lista) - 1`.

Por ejemplo, para recorrer la lista de ingresos por índices hacemos:

```python
for i in range(len(ingresos_mensuales)):
    print("Ingreso en posición", i, ":", ingresos_mensuales[i])
```

En este caso, `len(ingresos_mensuales)` devuelve 12, que es la longitud de la lista. Luego, `range(12)` genera la secuencia de enteros desde 0 hasta 11 que se corresponden con los índices de la lista.

También podemos calcular la longitud de la lista antes del ciclo y guardarla en una variable para mayor claridad:

```python
n = len(ingresos_mensuales)
for i in range(n):
    print("Ingreso en posición", i, ":", ingresos_mensuales[i])
```

### 🧰 Métodos de listas

Las listas en Python tienen **métodos**, que son funciones incorporadas que nos permiten hacer diferentes operaciones.

Para usar un método de lista empleamos la sintaxis de **notación punto**:

```python
lista.metodo(argumentos)
```

En esta sintaxis:
- `lista` es el nombre de la lista sobre la que queremos operar,
- `metodo` es el nombre del método,
- y los paréntesis contienen los **argumentos** que el método necesita (a veces no necesita ninguno).

🔍 Si querés conocer todos los métodos de listas disponibles, saber qué hace cada uno y cómo se usa correctamente, consultá la [documentación oficial de listas](https://docs.python.org/es/3.13/tutorial/datastructures.html) de Python.

En el ejemplo anterior trabajamos con tres métodos muy comunes:
- `lista.index(x)`
   
    Busca y devuelve el **índice** de la **primera aparición** del elemento `x` en la lista.

    Si el valor no está, **lanza un error** (`ValueError`).
- `lista.append(x)`

    Agrega el elemento `x` al **final** de la lista.
    
    No devuelve una lista nueva, simplemente modifica la original.
- `lista.sort()`

    **Ordena** los elementos de la lista en orden creciente.

    También modifica la lista original.

👉 A las operaciones que modifican directamente la lista sin crear una nueva se las llama modificaciones **in place** (en el lugar).

**🔄 Comparación con formas anteriores**

El caso de `append()` vale la pena considerarlo ya que antes, para **agregar elementos a una lista** usábamos **concatenación y reasignación**:



```python
lista = lista + [nuevo_elemento]
```

Ahora, con `append()`, podemos hacer lo mismo de forma más directa:

```python
lista.append(nuevo_elemento)
```

Por otro lado, los métodos `index()` y `sort()`, encapsulan tareas que antes resolvimos con nuestros propios algoritmos, como:

- Buscar un valor en una lista (algoritmo de *búsqueda lineal o binaria*),
- Ordenar una lista (algoritmo de *ordenamiento por selección*, por ejemplo).

📌 **La idea no es olvidar esos algoritmos**, ya que son los que están implementados detrás de estos métodos, sino más bien entender que ahora que los conocemos, podemos apoyarnos en estas herramientas para resolver nuevos desafíos.

✅ **Métodos in-place (modifican la lista original)**

Estos métodos **no devuelven una lista nueva**, sino que cambian el contenido de la lista sobre la que se aplican.

Algunos ejemplos son:
- `lista.append(x)`: agrega un elemento al final de la lista.
- `lista.sort()`: ordena la lista.
- `lista.reverse()`: invierte el orden de los elementos.
- `lista.extend(otra_lista)`: agrega los elementos de `otra_lista` al final de la lista.
- `lista.insert(i, x)`: inserta el elemento `x` en la posición `i`.
- `lista.remove(x)`: elimina la primera aparición del elemento `x`.

📌 Como estos métodos no devuelven una lista nueva, no tiene sentido hacer algo como `nueva_lista = lista.append(x)` porque `nueva_lista` quedaría como `None`.

📌 `None` es un valor especial en Python que representa “nada” o “ausencia de valor”.

🔁 **Métodos que devuelven una nueva lista**

Estos métodos **no modifican la lista original**, sino que producen una **lista nueva** con el resultado de la operación.

Un ejemplo de esto es:
- `lista.copy()`: crea una copia de la lista.

📌 Es imporante recordar que operaciones como `lista + [x]` o `lista[1:]` (rebanado) también crean nuevas listas.

📌 Estos métodos son útiles si querés **preservar la lista original** y trabajar con una versión modificada.

## 📦 Dividiendo los datos en partes

En clases pasadas aprendimos que la **mediana** nos ayuda a ubicar el "centro" de un conjunto de datos.

La mediana es el valor que deja el 50% de los valores por debajo y al otro 50% por encima.

Ahora nos preguntamos 🤔:

- ¿Podemos ir más allá y dividir los datos en más partes iguales?
- ¿Existen otros valores además de la mediana que nos ayuden a entender cómo se distribuyen los datos?

Fijate este código y analizá cada línea. Luego ejecutalo y observá los resultados:

````python
ingresos_mensuales = [570, 1200, 590, 500, 600, 1500, 610, 550, 630, 650, 620, 580]

datos = ingresos_mensuales.copy()
datos.sort()

n = len(datos)

# Q2 (mediana)
if n % 2 == 0:
    Q2 = (datos[n // 2 - 1] + datos[n // 2]) / 2
else:
    Q2 = datos[n // 2]

if n % 2 == 0:
    mitad_inferior = datos[:n // 2]
    mitad_superior = datos[n // 2:]
else:
    mitad_inferior = datos[:n // 2]
    mitad_superior = datos[n // 2 + 1:]

# Q1 (mediana de la mitad inferior)
n1 = len(mitad_inferior)
if n1 % 2 == 0:
    Q1 = (mitad_inferior[n1 // 2 - 1] + mitad_inferior[n1 // 2]) / 2
else:
    Q1 = mitad_inferior[n1 // 2]

# Q3 (mediana de la mitad superior)
n2 = len(mitad_superior)
if n2 % 2 == 0:
    Q3 = (mitad_superior[n2 // 2 - 1] + mitad_superior[n2 // 2]) / 2
else:
    Q3 = mitad_superior[n2 // 2]
````

````python
print("Ingresos ordenados:", datos)
print("Mitad inferior:", mitad_inferior)
print("Mitad superior:", mitad_superior)
````

````python
print("Q1:", Q1)
print("Q2:", Q2)
print("Q3:", Q3)
````

### 📊 Cuartiles

Los valores Q1, Q2 y Q3 que calculamos en el ejemplo anterior se conocen como **cuartiles**.

Un conjunto de datos ordenado puede dividirse en **cuatro partes iguales**, cada una con el **25%** de los datos.

Los cuartiles marcan los **límites** entre esas partes:

- **Q1 (primer cuartil)**: es el valor que deja al **25% de los datos por debajo** y al **75% por encima**.
- **Q2 (segundo cuartil o mediana)**: divide al conjunto a la mitad; el **50% de los datos están por debajo** y el otro **50% por encima**.
- **Q3 (tercer cuartil)**: es el valor que deja al **75% de los datos por debajo** y al **25% restante por encima**.

<div style="display: flex; justify-content: space-between;"> <img src="https://drive.google.com/uc?id=18xj-PQKRfYL3o2tCIAKouHiK3pKMcDFM" width="700" />

📌  Esto quiere decir que la **mitad central de los datos (el 50%) está entre Q1 y Q3**.

Para calcular los cuartiles seguimos este **algoritmo**:
1. **Ordenar** los datos de menor a mayor.
2. Calcular la **mediana (Q2)**.
3. Dividir el conjunto en **dos mitades**: mitad inferior y mitad superior.
4. Calcular la **mediana de la mitad inferior (Q1)** y la **mediana de la mitad superior (Q3)**.

En el ejemplo, usamos `datos = ingresos_mensuales.copy()` para **no modificar los datos originales**. Luego, aplicamos `datos.sort()` para ordenar la copia **in place** (en su lugar). Después, se calculan las tres medidas (Q1, Q2 y Q3) siguiendo los pasos anteriores, teniendo en cuenta si la cantidad de datos es par o impar.

📌 **¿Por qué son útiles los cuartiles?**

Los cuartiles son muy útiles porque nos permiten entender la **dispersión** y la **forma** de la distribución, saber **dónde se concentran los datos** (por ejemplo, si están todos cerca del centro o muy dispersos) y también identificar **valores atípicos** (*outliers*) cuando un dato está mucho más allá de Q1 o Q3, como ya veremos.

En el ejemplo de los ingresos mensuales, obtuvimos los siguientes cuartiles:
- Q1 = 575
- Q2 = 605
- Q3 = 640

Esto significa que:
- El 25% de las personas gana menos de 575 mil pesos.
- El 50% central de los ingresos está entre 575 y 640 mil pesos, es decir, en un rango de 65 mil pesos.
- El 75% de las personas gana menos de 640 mil pesos.

Al analizar la simetría de la distribución, podemos comparar los tramos:
- Q2 - Q1 = 605 - 575 = 30
- Q3 - Q2 = 640 - 605 = 35

Como el intervalo entre Q2 y Q3 es más amplio que entre Q1 y Q2, esto sugiere que hay ingresos más altos que se alejan del centro, **alargando la distribución hacia la derecha**. En otras palabras, la distribución se dice que es **asimétrica a la derecha**.

Si fuera el caso opuesto, se diría que la distribución es **asimétrica a la izquierda**. Y si por el contrario, los intervalos fueran iguales o muy similares, diríamos que la distribución es **simétrica**.

## 🕵️ ¿Y estos valores tan distintos?

Hasta ahora, aprendimos cómo dividir un conjunto de datos en partes iguales usando los cuartiles. Pero ahora queremos ir más allá.

¿Podemos usar los cuartiles para detectar valores que se alejan mucho del resto de los dato? 🤔

Mirá el siguiente código, pensá qué hace cada línea, y luego ejecutalo para ver los resultados:

````python
IQR = Q3 - Q1
LI = Q1 - 1.5 * IQR
LS = Q3 + 1.5 * IQR

outliers = []
for x in datos:
    if x < LI or x > LS:
        outliers.append(x)
````

````python
print("Ingresos ordenados:", datos)
print("IQR:", IQR)
print("Límite inferior:", LI)
print("Límite superior:", LS)
print("Datos atípicos:", outliers)
````

### ⚠️ Datos atípicos (outliers)

En datos reales es común encontrar algunos valores que se alejan mucho del patrón general.

A estos valores se los llama **datos atípicos** (*outliers*, en inglés).

Los datos atípicos son importantes porque pueden indicar:
- 🔍 **Casos excepcionales** (como ingresos muy altos o bajos),
- 🛠️ **Errores de carga o medición**,
- 🧩 **Personas con características muy distintas** al resto.

En nuestro conjunto de datos, los ingresos de 1.200 y 1.500 (es decir, 1.2 y 1.5 millones de pesos) son claramente **valores atípicos**, ya que están **muy alejados** del resto de los ingresos observados.



### ↔️ Rango intercuartil

Para analizar **cuán dispersos** están los datos, es decir, si están más bien agrupados o si varían mucho, usamos distintas medidas.

Dos de las más comunes que vimos son el rango y el desvío estándar.

Ahora sumamos una nueva: el **rango intercuartil o IQR** (*interquartile range*, en inglés). También llamado **distancia intercuartil** en libros.

El IQR es una medida de dispersión que mide la **longitud del intervalo del 50% central** de los datos.

Cuanto mayor sea el IQR, más dispersos están los datos en esa zona central.

Se calcula así:

$$IQR = Q3 - Q1$$

Es decir, es la diferencia entre:
- el **tercer cuartil (Q3)**: la mediana de la mitad superior,
- y el **primer cuartil (Q1)**: la mediana de la mitad inferior.

Para calcular el rango intercuartil aplicamos este **algoritmo**:
1. **Ordenar** los datos de menor a mayor.
2. Calcular la **mediana (Q2)**.
3. Calcular la **mediana de la mitad inferior (Q1)**.
4. Calcular la **mediana de la mitad superior (Q3)**.
5. Calcular el **IQR** haciendo Q3 - Q1.

📌 **¿Por qué es útil el IQR?**

Al igual que la mediana, el **rango intercuartílico (IQR) no se ve tan afectado por valores extremos o atípicos**.

Por eso, se lo prefiere como medida de **dispersión** cuando se sospecha que puede haber datos que distorsionen otras medidas de dispersión, como el **rango** o el **desvío estándar**.

En resumen:
- El **rango** mide la dispersión tomando el **máximo** y el **mínimo** valor del conjunto.
- El **desvío estándar** mide la dispersión de los datos con respecto a la **media** del conjunto.
- El **IQR** mide la dispersión de los datos centrales en torno a la **mediana** del conjunto.

📌 El **IQR es menos sensible** a los valores atípicos que el rango y el desvío estándar.

En el ejemplo de los ingresos vimos que Q1 = 675 y Q3 = 740, entonces IQR = 740 - 675 = 60.

Esto significa que la mitad central de los ingresos se encuentra en un intervalo de 65 mil pesos alrededor de la mediana (Q2 = 705).

🔍 **La regla del IQR para detectar outliers**

El IQR no solo sirve para medir cuán dispersos están los datos: también nos ayuda a **detectar valores atípicos**.

A este método se lo conoce como la **regla del IQR**.

La idea es definir un **rango razonable** para los datos: todo valor que quede **muy lejos** de ese rango será considerado **atípico**.

Para aplicar la regla del IQR, seguimos estos pasos:
1. Calcular el **IQR** de los datos:
$$IQR = Q3 - Q1$$
2. Multiplicar el **IQR por 1.5** (una constante que representa un margen de seguridad).
3. Sumar 1.5 x IQR al tercer cuartil (Q3). Este es el **límite superior (LS)**:
$$LS=Q3 + 1.5 \times IQR$$

4. Restar 1.5 x IQR al primer cuartil (Q1). Este es el **límite inferior (LI)**:
$$LI=Q1 - 1.5 \times IQR$$

🎯 Cualquier valor que esté **por encima de LS** o **por debajo de LI** se considera un **dato atípico** (outlier).

En el ejemplo de los ingresos obtuvimos que:
- Límite inferior (LI): 477.5
- Límite superior (LS): 737.5

En este caso, no hay ingresos en el conjunto más bajos que LI, pero sí hay dos valores más altos que LS, que son 1.200 y 1.500.

Estos dos ingresos son considerados *outliers* en nuestro conjunto.

📌 La regla del IQR es una **regla empírica**. No garantiza que todos los valores detectados sean incorrectos, pero sirve como una señal de advertencia para revisarlos con más atención.

### 🧮  Los cinco números resumen

Ahora que aprendimos los cuartiles tenemos todo lo necesario para poder describir nuestros datos.

Cualquier conjunto de datos se puede resumir con cinco valores, llamados **los cinco números resumen**.

Estos cinco números consisten en (en orden ascendente):

- **El valor mínimo** o más bajo del conjunto de datos.
- **El primer cuartil (Q1, 25%)**, que representa una cuarta parte de los datos.
- **La mediana (Q2, 50%)**, que representa el punto medio de los datos
- **El tercer cuartil (Q3, 75%)**, que representa tres cuartas partes de los datos
- **El valor máximo** o más alto del conjunto de datos.

Tomemos el ejemplo de los ingresos que vimos antes:

$$
\begin{array}{c c c c c c}
    & \text{Mínimo} & \text{Primer cuartil (Q1)} & \text{Mediana (Q2)} & \text{Tercer cuartil (Q3)} & \text{Máximo} \\
    & 500 & 575 & 605 & 640 & 1500
\end{array}
$$

🔎 ¿Qué podemos interpretar de estos valores?
- El 50% central de los ingresos está entre 575 y 640 mil pesos.
- El rango intercuartílico (IQR) de esta zona central es de 65 mil.
- El 25% inferior de las personas gana entre 500 y 575 mil.
- El 25% superior gana entre 640 mil y 1.5 millones de pesos.

El 25 % superior de los ingresos está mucho más disperso que el 25% inferior.

Esto sugiere que existen valores extremos muy altos que podrían estar alargando la distribución hacia la derecha (asimetría a la derecha).
