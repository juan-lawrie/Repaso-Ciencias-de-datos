En la fabricación de vigas de acero se utiliza una aleación cuya resistencia se evalúa en función de la **cantidad máxima de carga** (en toneladas) que una viga puede soportar sin romperse. Al parecer, la media de la muestra contradice lo que afirma la empresa (3.85 toneladas).

Sin embargo, la persona encargada de recolectar los datos cometió algunos errores de tipeo que resultaron en **datos atípicos en la muestra** y pueden estar afectando el valor de la media.

Para averiguar cuales son los valores atípicos, debajo disponemos de una función que detecta y devuelve los datos atípicos usando la regla del IQR:

```python
def calcular_mediana(lista, ordenada=False):
    if ordenada:
        lista_ordenada = lista.copy()
    else:
        lista_ordenada = sorted(lista)

    n = len(lista)
    mitad = n // 2

    if n % 2 == 0:
        mediana = (lista_ordenada[mitad-1] + lista_ordenada[mitad]) / 2
    else:
        mediana = lista_ordenada[mitad]

    return mediana

def calcular_cuartiles(lista):
    lista_ordenada = sorted(lista)

    n = len(lista_ordenada)
    mitad = n // 2

    if n % 2 == 0:
        mitad_inferior = lista_ordenada[:mitad]
        mitad_superior = lista_ordenada[mitad:]
    else:
        mitad_inferior = lista_ordenada[:mitad]
        mitad_superior = lista_ordenada[mitad + 1:]

    Q1 = calcular_mediana(mitad_inferior, ordenada=True)
    Q2 = calcular_mediana(lista_ordenada, ordenada=True)
    Q3 = calcular_mediana(mitad_superior, ordenada=True)

    return Q1, Q2, Q3

def detectar_atipicos(lista):
    Q1, Q2, Q3 = calcular_cuartiles(lista)

    IQR = Q3 - Q1
    LI = Q1 - 1.5 * IQR
    LS = Q3 + 1.5 * IQR

    atipicos = []

    for dato in lista:
        if dato < LI or dato > LS:
            atipicos.append(dato)

    return atipicos
```

Usá esta función para ver si existen datos atípicos en la muestra de resistencias:

```python
atipicos = detectar_atipicos(resistencias)
print("Atipicos:", atipicos)
```

Si todo salió bien, habrás encontrado 3 datos atípicos en la muestra.

Lo que haremos entonces es **quitarlos de la muestra** y volver a calcular la media.

Completá el código debajo que armar una lista con los valores de resistencia sin los datos atípicos:

```python
a,b,c = atipicos
resistencias_sin_atipicos = [x for x in resistencias if (x!= a) and (x!= b) and (x!= c)]
#for ...
   # if r not in ...
     #   resistencias_sin_atipicos....

print("Tamaño de la muestra sin atípicos:", len(resistencias_sin_atipicos))
```

Ahora con esta muestra volvé a calcular la resistencia media:

```python
media_muestra_sin_atipicos = calcular_media(resistencias_sin_atipicos)
print("Media de la muestra sin atípicos:", round(media_muestra_sin_atipicos, 2))
```
