## 🎨 Encuesta: ¿color favorito?

Contamos con los resultados de una encuesta en la que 12 personas indicaron su color favorito.

Ejecutá el código para ver las respuestas:

````python
respuestas = [
    "rojo", "azul", "verde", "rojo", "azul", " azul ",
    "Rojo", "azul ", "Verde", "verde", "ROJO", "AZUL"
]

print("Respuestas originales:")
print(respuestas)
````

Como notarás, algunas respuestas tienen espacios innecesarios o están escritas con mayúsculas/minúsculas mezcladas.

Por lo tanto, lo que haremos será **limpiar las respuestas** para poder trabajar mejor.

Mirá este código y analizá cada línea del mismo:

````python
respuestas_limpio = []

for r in respuestas:
    r_limpio = r.strip()
    r_limpio = r_limpio.lower()
    respuestas_limpio.append(r_limpio)

print("Respuestas limpias:")
print(respuestas_limpio)
````

2. ¿Qué creés que hacen `.strip()` y `.lower()` en cada respuesta? Podés usar `print()` dentro del ciclo para verificarlo.
3. ¿Se pueden aplicar ambas operaciones en una única línea? Probá hacer `r_limpio = r.strip().lower()`.

Ahora queremos **resumir los datos numéricamente**, así que calcularemos las frecuencias de cada color:

````python
frecuencias = {}

for r in respuestas_limpio:
    if r not in frecuencias:
        frecuencias[r] = 1
    else:
        frecuencias[r] += 1

print(frecuencias)
````

### 🧰 Métodos de cadenas

Al igual que las listas, las **cadenas** también tienen métodos.

Los **métodos de cadenas** son funciones que nos permiten hacer operaciones sobre las cadenas.

Estas operaciones incluyen limpiarlas, transformarlas o extraer partes.

Para usar los métodos, aplicamos la sintaxis de **notación punto** como con los métodos de listas:

```python
cadena.metodo(argumentos)
```

En esta sintaxis:
- `cadena` es el nombre de la cadena sobre la que queremos operar,
- `metodo` es el nombre del método que queremos usar,
- y los paréntesis contienen los **argumentos** que el método necesita (a veces no necesita ninguno).

Dos métodos muy útiles que vimos en el ejemplo son:
- `.strip()`: elimina los espacios en blanco al principio y al final de la cadena.
- `.lower()`: convierte todas las letras a minúsculas.

📌 Estos métodos son útiles para **normalizar los datos**, es decir, dejarlos en un formato uniforme antes de analizarlos.

🔍 Si querés conocer todos los **métodos de cadenas** disponibles, consultá la [documentación oficial de cadenas](https://docs.python.org/es/3.13/library/stdtypes.html#string-methods) de Python.

### 🗂️ Diccionarios

Un **diccionario** es una estructura de datos que guarda información en **pares** de tipo **clave y valor**.

Pensemos en un diccionario de palabras: la clave sería la palabra y el valor sería su significado.

En programación funciona igual: **cada clave tiene un valor asociado**.

En el ejemplo que vimos tenemos los colores elegidos por algunas personas.

Para contar cuántas veces se eligió cada color (es decir, la frecuencia de cada color) usamos un diccionario:

```python
frecuencias = {'rojo': 4, 'azul': 5, 'verde': 3}
```

En la compu, podemos imaginar un diccionario representado así:

<div style="display: flex; justify-content: space-between;"> <img src="https://drive.google.com/uc?id=1JcmdfDlFzeDwZQ1W9A9Zey1DdBH8qRS6" width="300" />

En este caso, `frecuencias` es la variable que refiere a un diccionario en la memoria que contiene **3 pares clave-valor**:
- Claves: `"rojo"`, `"azul"` y `"verde"` (las categorías)
- Valores: `4`, `5` y `3` (las cantidades o frecuencias de cada color)

Para **acceder a un valor** del diccionario, escribimos el nombre de la clave entre corchetes `[]`:

```python
print(frecuencias["azul"])  # Imprime: 5
```

El operador `in`, que ya usamos con listas, también funciona con diccionarios.

En este caso, se utiliza para **verificar si existe una clave** en el diccionario:



```python
print("rojo" in frecuencias)    # Imprime: True
print("amarillo" in frecuencias)    # Imprime: False
```

Esto nos sirve en el ejemplo para **inicializar** en 1 la frecuencia de una categoría nueva o **incrementar** en 1 la frecuencia si ya existe.


🔧 **Métodos útiles de diccionarios**

Los diccionarios también tienen **métodos** que nos permiten trabajar con ellos:

Algunos de los métodos más utiles son:
- `.keys()`→ devuelve solo las claves
- `.values()` → devuelve solo los valores
- `.items()`→ devuelve las parejas clave:valor


Por ejemplo:

```python
print(frecuencias.keys())   # dict_keys(['rojo', 'azul', 'verde'])
print(frecuencias.values()) # dict_values([4, 5, 3])
print(frecuencias.items())  # dict_items([('rojo', 4), ('azul', 5), ('verde', 3)])
```

Estos métodos devuelven una *vista* del diccionario, un tipo de dato un poco raro para nosotrxs.

Para transformar esta vista en una lista, podemos usar la función `list()`:

```python
print(list(frecuencias.keys()))   # ['rojo', 'azul', 'verde']
print(list(frecuencias.values())) # [4, 5, 3]
print(list(frecuencias.items()))  # [('rojo', 4), ('azul', 5), ('verde', 3)]
```

🔍 Si querés conocer todos los **métodos de diccionarios** disponibles, consultá la [documentación oficial de diccionarios](https://docs.python.org/es/3.7/tutorial/datastructures.html#dictionaries) de Python.

🔁 **Recorrer un diccionario**

Podemos recorrer un diccionario de varias maneras.

Una de ellas es **recorrer sólo las claves** usando el método `.keys()` y acceder al valor usando la clave:

```python
for color in frecuencias.keys():
    print(color, frecuencias[color])
```

La otra forma es **recorrer clave y valor al mismo tiempo** usando el método `.items()`:

```python
for color, cantidad in frecuencias.items():
    print(color, cantidad)
```

## 🧹 Limpiando las respuestas de la encuesta

Ahora contamos con otra encuesta donde hay más colores dentro de los elegidos:



````python
respuestas = [
    "rojo", "azul", "verde", "rojo", "azul", " azul ",
    "Rojo", "azul ", "Berde", "verde", "ROJO", "AZUL",
    "Amarillo ", "NEgro", "AMARILLO"
]

print("Respuestas originales:")
print(respuestas)
````

El código de abajo pretende limpiar los datos, como hicimos antes, y corregir algunos errores de ortografía.

Sin embargo, tiene algunos errores. ¿Te animás a corregirlo? 🤓

````python
respuestas_limpio = []

for r in respuestas:
    r_limpio = r.strip().lower()

    if r_limpio == "berde":
        r_limpio = "v" + r_limpio[1:]

    respuestas_limpio.append(r_limpio)

print("Respuestas limpias:")
print(respuestas_limpio)
````

Si ya corregiste el código y limpiaste los datos, calculá las frecuencias absolutas y relativas de cada color:

````python
frecuencias = {}

for r in respuestas_limpio:
    if r not  in frecuencias:
        frecuencias[r] = 1
    else:
        frecuencias[r] += 1

print("Frecuencias absolutas:")
print(frecuencias)
````

````python
total_respuestas = sum(frecuencias.values())

frecuencias_relativas = frecuencias.copy()

for color in frecuencias_relativas.keys():
    frecuencias_relativas[color] /= total_respuestas

print("Frecuencias relativas:")
print(frecuencias_relativas)
````

### 🔤 Cadenas como datos secuenciales

Las cadenas, al igual que las listas, son estructuras de **datos secuenciales**. Esto significa que:
- Podés **acceder a cada carácter por su posición** (indexación),
- Podés **recorrerlas con ciclos** (for o while),
- Y podés usar la función `len()` para **saber cuántos caracteres tienen**.

En el caso de la **indexación**, cada carácter tiene una posición (índice) que comienza en 0:

```python
color = "azul"
print(color[0])  # Imprime: a (primer caracter)
print(color[2])  # Imprime: u (tercer caracter)
```

También podemos usar **índices negativos** para contar desde el final:

```python
print(color[-1])  # Imprime: l (último caracter)
```

La función `len()` nos devuelve la **longitud** de la cadena:

```python
print(len(color))  # Imprime: 4
```

En cuanto al **recorrido** de una cadena, podemos hacerlo con un ciclo `for`, y como de costumbre, tenemos dos formas:


1.Directamente por los caracteres:
```python
for c in color:
    print(c)
```

2.Usando los índices:
```python
for i in range(len(color)):
    print(color[i])
```

Ambas formas imprimen los caracteres uno por uno:
```python
a
z
u
l
```

🔒 **Inmutabilidad de las cadenas**

Una diferencia clave entre cadenas y listas es que **las cadenas son inmutables**.
Esto significa que **no podés cambiar sus caracteres directamente**. En cambio, **cada vez que las querés modificar, necesitás crear una nueva cadena**.

Por ejemplo:

```python
color = "Azul"
color = color.lower()   # Creamos una nueva cadena en minúsculas
print(color)    # Imprime: azul
```

📌 Como las cadenas son inmutables, **ningún método modifica la cadena original "en el lugar"** (*in place*). Siempre devuelven una nueva versión.

### ✍️ Crear y modificar un diccionario

Al igual que las listas, **los diccionarios son mutables**, lo que significa que **podemos modificarlos después de crearlos**.

Podemos empezar con un diccionario vacío:

```python
frecuencias = {}
```

Y luego agregar un par de clave y valor así:

```python
frecuencias["rojo"] = 1   # Agrega la clave "rojo" con valor 1
```

Cuando esa clave ya existe, luego podemos modificar su valor:

```python
frecuencias["rojo"] += 1  # Suma 1 al valor actual
```

📌 Las **claves no se repiten**: si agregamos una clave que ya existe, estaremos cambiando su valor.

Para saber cuántos pares hay en un diccionario usamos `len()`:

```python
print(len(frecuencias))  # Imprime: 1 (porque hay un sólo par rojo: 1)
```

✅ **¿Qué pueden ser las claves y los valores?**

- Las **claves** deben ser **datos inmutables**: cadenas, números, booleanos.
- Los **valores** pueden ser **cualquier tipo de dato**: números, listas, otros diccionarios, etc.

## 📊 Visualización de datos cualitativos

Ahora queremos saber si hay una manera de **resumir** los resultados de la encuesta **de forma visual**.

Esto podría ayudarnos a entender mejor los datos y a compararlos más fácilmente que si solo miramos los números.

Fijate este código y pensá que hace cada línea:

````python
import matplotlib.pyplot as plt

categorias = list(frecuencias.keys())
cantidades = list(frecuencias.values())

plt.figure()
plt.bar(categorias, cantidades)
plt.title("Frecuencias de colores favoritos")
plt.show()
````

Ahora mirá este otro código:

````python
plt.figure()
plt.pie(cantidades, labels=categorias, autopct="%1.1f%%")
plt.title("Distribución de colores")
plt.axis("equal")
plt.show()
````

📌 Preguntas:
1. ¿Qué información muestra el primer gráfico? Las categorias y las frecuencias de colores favoritos. ¿Qué representa la altura de cada barra? Las frecuencias absolutas de los colores, rojo, verde y azul.
2. ¿Qué información muestra el segundo gráfico? La distribucion de los colores con respecto al total de categorias  ¿Qué representa el tamaño de cada porción de la torta? El tamaño total de cada porcion representa el porcentaje que representa cada categoria del total
3. ¿Qué ventajas y desventajas ves en cada gráfico? La ventaja del grafico de barras es que te ayuda a ver claramente las cantidades o frecuencias absolutas. Las barras hacen que sea mucho más facil comparar visualmente los valores entre las distintas categorías.
La otra ventaja es que si una categoria tiene un valor de cero, sigue apareciendo en el gráfico (con una barra de altura cero y su nombre en el eje horizontal), por lo que no perdes esa información.
Desventajas: No es la mejor opción si tu objetivo principal es mostrar rapidamente qué porcentaje del total general representa una categoría específica.

La ventaja del grafico de torta es que es ideal para ver proporciones relativas. Al estar dividido en porciones, hace que sea visualmente muy facil identificar qué parte del todo (el 100%) representa cada categoría.

Desventajas: Si tienes categorias con valores muy similares, es difícil comparar a simple vista cuál porción es más grande sin leer los números.
Si una categoria da cero, no se puede graficar porque un porcentaje del 0% no tiene área y no genera ninguna porción. Esto hace que la categoria desaparezca del circulo y usualmente se omita en la leyenda

¿Cuál te ayuda más a ver las cantidades y cuál las proporciones?El grafico de barras te ayuda a ver mas las cantidades y el de torta te ayuda a ver las proporciones.

### 📚 Bibliotecas

Cuando trabajamos con datos a veces necesitamos hacer **gráficos** para describir mejor la información que tenemos.

Pero si lo pensamos un ratito... enseñarle a la compu a **dibujar un gráfico** no es nada fácil: tendríamos que decirle cómo crear una hoja, cómo dibujar líneas, cómo poner colores, nombres, valores, etc. Sería muchísimo trabajo.

Por suerte, existen **herramientas ya hechas** que se encargan de todas esas cosas difíciles. A esas herramientas les llamamos **bibliotecas**.

Una biblioteca en programación es como una **caja de herramientas** que ya viene con muchas funciones listas para usar.

📈 **La biblioteca Matplotlib**

Para hacer gráficos con la compu, vamos a usar una biblioteca muy conocida que se llama **Matplotlib**.

Más específicamente, usaremos una parte de Matplotlib llamada `pyplot`, que está pensada para crear gráficos de manera sencilla.

A lo largo de la carrera vas a trabajar con muchas otras bibliotecas, por ejemplo:
- **NumPy**: para trabajar con arreglos y cálculos numéricos.
- **Pandas**: para organizar, transformar y analizar datos en tablas.
- **Scikit-learn**: para crear modelos de machine learning (aprendizaje automático).

🔍 Si querés conocer más sobre Matplotlib, podés consultar la [documentación oficial de Matplotlib](https://matplotlib.org/).

📦 **¿Cómo usamos una biblioteca?**

Antes de usar una biblioteca, tenemos que **importarla**. Esto le dice a la compu: “voy a usar esta biblioteca".

Para importar Matplotlib, escribimos:

```python
import matplotlib.pyplot as plt
```

¿Qué significa cada parte?
- `import`: la palabra clave para importar bibliotecas.
- `matplotlib`: el nombre de la biblioteca.
- `pyplot`: un módulo especial de Matplotlib para hacer gráficos.
- `as plt`: le damos un nombre más corto (alias), para no tener que escribir todo cada vez.

📌 Con un alias evitamos escribir `matplotlib.pyplot` cada vez que usamos una función. Solo usamos `plt`.

📌 Las bibliotecas **se importan una sola vez** al inicio del código, y luego ya están disponibles para usar en todas las celdas.

🔧 **¿Cómo se usan las funciones de una biblioteca?**

Una vez que importamos la biblioteca, podemos usar sus funciones con la **notación punto**:

```python
plt.funcion(argumentos)
```

En esta sintaxis:
- `plt` es la biblioteca (con el alias que le dimos),
- `funcion` es el nombre de la función que queremos usar,
- y dentro de los paréntesis van los **argumentos** que necesita (a veces no necesita ninguno).


Por ejemplo, algunas de las funciones que usamos fueron:
- `plt.figure()`: crea una figura donde se va a dibujar el gráfico (lienzo).
- `plt.bar()`: dibuja un gráfico de barras.
- `plt.pie()`: dibuja un gráfico de torta.
- `plt.title()`: agrega un título al gráfico.
- `plt.axis("equal")`: hace que el gráfico se vea cuadrado.
- `plt.xlabel()`: agrega una etiqueta al eje X.
- `plt.ylabel()`: agrega una etiqueta al eje Y.
- `plt.show()`: muestra el gráfico en pantalla.

🖼 **Crear un gráfico paso a paso**

Para hacer un gráfico, seguimos estos **pasos en orden**:

```python
plt.figure()        # Crea una figura o "lienzo"
plt.bar(...)        # O plt.pie(...) según el tipo de gráfico
plt.title(...)      # Título del gráfico
plt.xlabel(...)     # Opcional: nombre del eje X
plt.ylabel(...)     # Opcional: nombre del eje Y
plt.show()          # Muestra el gráfico en pantalla
```
En resumen, primero creamos la figura.

Luego, dibujamos el gráfico (barras, torta, línea, etc.).

Despúes, personalizamos el gráfico (le agregamos título, etiqueta a los ejes, etc.).

Finalmente, mostramos el gráfico resultante.

📌 El orden de estas instrucciones importa:
- Si no usamos `plt.figure()` al principio, no tenemos dónde dibujar.
- Si no usamos `plt.show()`, el gráfico no aparece.

### 📊 Gráficos para datos cualitativos

Cuando nuestros datos son cualitativos podemos usar dos tipos de gráficos: de **barras** y de **torta**.

🟦 **Gráfico de barras**

Muestra **cantidades** (frecuencias) para cada categoría:
- Las **categorías** se ven en el eje horizontal.
- Las **frecuencias** se ven en el eje vertical.
- La **altura de cada barra** representa cuántas veces aparece cada categoría.

💡 Es útil para **comparar cantidades** entre categorías.

🍰 **Gráfico de torta**

Muestra **proporciones** (qué parte representa cada categoría del total):
- Cada **porción** de la torta representa una **categoría**.
- El tamaño de cada porción muestra **cuánto representa** esa categoría.

💡 Es útil para ver **cómo se divide un todo**.

## 🎼 Encuesta: ¿género músical favorito?

Supongamos ahora que se hizo una encuesta a 30 personas sobre música preferida y estos fueron sus resultados:

````python
respuestas = [
    "Rock", "pop", "trap", "rock", "cumbia", "Rock", "Pop", "trap", "Trap",
    "cumbia", "Clásica", "cumbia", "rock", "pop", "trap", "Rock", "Pop",
    "Trap", "clásica", "rock", "CUMBIA", "Trap", "Pop", "rock", "Clásica",
    "trap", "pop", "ROCK", "cumbia", "rock"
]

print("Respuestas originales:")
print(respuestas)
````

Vamos a limpiar los datos y a calcular luego las frecuencias absolutas de cada opción:

````python
respuestas_limpio = []

for r in respuestas:
    r_limpio = r.strip().lower()
    respuestas_limpio.append(r_limpio)

print("Respuestas limpias:")
print(respuestas_limpio)
````

````python
frecuencias = {}

for r in respuestas_limpio:
    if r not in frecuencias:
        frecuencias[r] = 1
    else:
        frecuencias[r] += 1

print("Frecuencias absolutas:")
print(frecuencias)
````

Ahora hacé el gráfico de barras y de torta para este caso:

````python
import matplotlib.pyplot as plt
categorias = list(frecuencias.keys())
cantidades = list(frecuencias.values())

plt.figure()
plt.bar(categorias, cantidades)
plt.title("Frecuencias de géneros musicales favoritos")
plt.show()
````

````python
plt.figure()
plt.pie(cantidades, labels=categorias, autopct="%1.1f%%")
plt.title("Distribución de géneros musicales")
plt.axis("equal")
plt.show()
````

### 💻 Gráficos de barras y torta con Matplotlib

🟦 **Gráfico de barras**

Para hacer este gráfico escribimos lo siguiente:

```python
plt.bar(categorias, cantidades)
```
La función recibe varios argumentos pero por ahora sólo nos interesan los dos primeros:
- `categorias` indica los **nombres de cada categoría** (uno por cada valor).
- `cantidades`: es la lista con las **frecuencias** o cantidades que queremos representar.

📌 Importante:
- Las listas `categorias` y `cantidades` deben tener la **misma cantidad de elementos**.
- Deben pasarse **respetando el orden**: primero las categorias y luego las frecuencias.

🍰 **Gráfico de torta**

Para hacer este gráfico escribimos lo siguiente:

```python
plt.pie(cantidades, labels=categorias, autopct="%1.1f%%")
```
En este caso:
- `cantidades`: es la lista con las **frecuencias** o cantidades que queremos representar.
- `labels=categorias`: indica los **nombres de cada categoría** (uno por cada valor).
- `autopct="%1.1f%%"`: le indica a la compu que queremos **mostrar los porcentajes** en el gráfico, con un **decimal** de precisión.

📌 Importante:
- Las listas `categorias` y `cantidades` deben tener la misma cantidad de elementos.
- No olvidar escribir los parámetros `labels=` y `autopct=`, ya que si se omiten el gráfico no mostrará los nombres ni los porcentajes.
