Para comenzar el análisis, queremos saber **cuántas propiedades tienen o no tienen vista al golfo**.

Para ello, vamos a definir una función que calcule las frecuencias absolutas de un conjunto de datos cualitativos:

````python
def calcular_frecuencias(lista):
    frecuencias = {}

    for x in lista:
        if x not in frecuencias:
            frecuencias[x] = 1
        else:
            frecuencias[x] += 1

    return frecuencias
````

Luego, usamos esta función para obtener las frecuencias de vista de las propiedades:

````python
frecuencias_vista = calcular_frecuencias(vista_propiedad)
print("Frecuencias de tipo de vista:", frecuencias_vista)
````

````python
frecuencias_venta = calcular_frecuencias(tipo_venta)
print("Frecuencias de tipo de venta:", frecuencias_venta)
````

Sin embargo, analizar las frecuencias por separado no nos sirve de mucho. 🤔

- **¿Cuántas propiedades sin vista al golfo se vendieron de forma rápida, media o dilatada?**

Para responder estas preguntas necesitamos una forma de **resumir la información combinada de vista y rapidez de venta**.

````python
tabla_contingencia_absoluta = {}

for i in range(len(vista_propiedad)):
    vista = vista_propiedad[i]
    tipo = tipo_venta[i]

    if vista not in tabla_contingencia_absoluta:
        tabla_contingencia_absoluta[vista] = {}

    if tipo not in tabla_contingencia_absoluta[vista]:
        tabla_contingencia_absoluta[vista][tipo] = 1
    else:
        tabla_contingencia_absoluta[vista][tipo] += 1

for vista, conteos in tabla_contingencia_absoluta.items():
    print(vista + ":", conteos)
````

### 🧮 Tabla de contingencia

Una **tabla de contingencia** es una herramienta usada en estadística para resumir la **asociación entre dos conjuntos de datos cualitativos**.

Es una **tabla de doble entrada** con las siguientes características:
- **Filas**: muestran las diferentes categorías de un conjunto de datos.
- **Columnas**: muestran las categorías del otro conjunto.
- **Celdas**: indican el número de casos (frecuencia) que combinan ambas categorías.

En el ejemplo que vimos la tabla de contingencia nos quedó así:

            |                | Rápida | Media | Dilatada | TOTAL |
            |----------------|--------|-------|----------|-------|
            | Golfo          | 9      | 26    | 5        | 40    |
            | NoGolfo        | 5      | 9     | 4        | 18    |
            |----------------|--------|-------|----------|-------|
            | TOTAL          | 14     | 35    | 9        | 58    |


De esta tabla podemos interpretar, por ejemplo, que:
- Entre las propiedades **con vista al golfo**, se vendieron 9 de forma rápida, 26 de forma media y 5 de forma dilatada.
- Entre las propiedades **sin vista al golfo**, se vendieron 5 de forma rápida, 9 de forma media y 4 de forma dilatada.

Del mismo modo, la cantidad de propiedades sin vista al golfo que se vendieron de forma rápida, media y dilatada fue 5, 9 y 4, respectivamente.

También podemos ver los **totales por fila y por columna**, los cuales representan las **frecuencias individuales** de las categorías:
- Al final de cada fila está el total de propiedades según su vista: 40 con vista al golfo y 18 sin vista al golfo.
- Al final de cada columna está el total según la rapidez de venta: 14 rápidas, 25 medias y 9 dilatadas.

La suma de todos los valores es 58, el total de datos que disponemos: 40 + 18 = 58 y 14 + 25 + 9 = 58.

### 💻 Representación de la tabla con diccionarios anidados

Para guardar la tabla de contingencia en la compu usamos usamos un diccionario *dentro* de otro diccionario.

A esto se le llama **diccionario anidado**.

La tabla que vimos en el ejemplo:

            |                | Rápida | Media | Dilatada |
            |----------------|--------|-------|----------|
            | Golfo          | 9      | 26    | 5        |
            | NoGolfo        | 5      | 9     | 4        |
            |----------------|--------|-------|----------|

Se guarda así:

```python
Golfo: {'Rápida': 9, 'Media': 26, 'Dilatada': 5}
NoGolfo: {'Rápida': 5, 'Media': 9, 'Dilatada': 4}
```

En este diccionario, los pares son las filas de la tabla:
- Cada **clave principal** ("Golfo" o "NoGolfo") representa una **categoría del primer conjunto de datos** (tipo de vista).
- El **valor asociado** es otro diccionario, donde cada clave representa una **categoría del segundo conjunto** (rapidez de venta), y el valor indica **cuántas veces ocurre esa combinación**.

🛠 **Cómo se construye paso a paso**

Para calcular la **tabla de contingencia de frecuencias absolutas** recorremos los datos y hacemos lo siguiente:

- Si la clave principal (por ejemplo "Golfo") **no está en el diccionario**, la agregamos y le asignamos un diccionario vacío:

```python
if vista not in tabla_contingencia_absoluta:
    tabla_contingencia_absoluta[vista] = {}
```

- Luego actualizamos la cuenta (frecuencia) de la combinación. Si es la primera vez que aparece, empezamos con 1. Si ya está, sumamos 1:

```python
if vista not in tabla_contingencia_absoluta:
        tabla_contingencia_absoluta[vista] = {}

    if tipo not in tabla_contingencia_absoluta[vista]:
        tabla_contingencia_absoluta[vista][tipo] = 1
    else:
        tabla_contingencia_absoluta[vista][tipo] += 1
```

Cada combinación de categorías (por ejemplo, "Golfo" y "Rápida") se guarda en una estructura que permite contar **cuántas veces aparece**.

El diccionario principal guarda las **filas** de la tabla, y cada diccionario interno guarda los valores de esa fila.

En el código también construímos la **tabla de contingencia de frecuencias relativas**.

Para ello, recorremos cada fila de la tabla de frecuencias absolutas y hacemos lo siguiente:

- Agregamos la clave principal (por ejemplo "Golfo") al nuevo diccionario y le asignamos un diccionario vacío:

```python
tabla_contingencia_relativa[vista] = {}
```
- Luego, para cada categoría de tipo de venta ("Rápida, "Media", "Dilatada"), agregamos la frecuencia relativa al total de datos:

```python
for tipo, cantidad in conteos_tipo.items():
    tabla_contingencia_relativa[vista][tipo] = cantidad / total_datos
```

En esta tabla, cada combinación de categorías (por ejemplo, "Golfo" y "Rápida") ahora muestra la **proporción de veces que aparece esa combinación con respecto al total de datos**.

Es decir, no cuenta cuántas veces ocurre, sino **qué parte del total representa** (un valor entre 0 y 1).

```python
Golfo: {'Rápida': 9, 'Media': 26, 'Dilatada': 5}
NoGolfo: {'Rápida': 5, 'Media': 9, 'Dilatada': 4}
```

Si miramos sólo la categoría *Rápida*, vemos que:
- **9 propiedades con vista al golfo** se vendieron rápido.
- **5 propiedades sin vista al golfo** se vendieron rápido.

A simple vista, podría parecer que las casas con vista al golfo se venden más rápido.

Pero para comparar correctamente, necesitamos tener en cuenta **el total de casas en cada grupo**.

Es decir, **no alcanza con comparar 9 vs 5**.

Debemos calcular **qué proporción del total representa cada valor**.



Para eso, construiremos una **tabla de frecuencias relativas condicionadas por vista**, que nos muestre los **porcentajes dentro de cada vista**.

Mirá el código que sigue y tratá de entender cada línea antes de ejecutarlo:

````python
tabla_contingencia_condicional = {}

for vista, conteos_tipo in tabla_contingencia_absoluta.items():
    total_fila = sum(conteos_tipo.values())

    tabla_contingencia_condicional[vista] = {}
    for tipo, cantidad in conteos_tipo.items():
        tabla_contingencia_condicional[vista][tipo] = cantidad / total_fila

for vista, proporciones_tipo in tabla_contingencia_condicional.items():
    print(vista + ":", proporciones_tipo)
````

### 🧮 Tabla de contingencia condicional

Cuando relacionamos dos conjuntos de datos cualitativos debemos tener cuidado sobre cuál de ellos calculamos las frecuencias relativas, es decir, cuáles son los conjuntos de datos que queremos comparar en esta relación.

En nuestro ejemplo, queremos ver cómo la **vista** influye en la **rapidez de venta** de una propiedad, entonces las relativas deben calcularse sobre las categorías de **vista**

Para ello, vamos a construir una **tabla de frecuencias relativas condicionadas por tipo de vista**.

Eso significa que, dentro de cada grupo de vista, calculamos **el porcentaje de propiedades** que se vendieron en cada categoría de rapidez.

Esto se hace **dividiendo cada valor por el total de su fila** (es decir, dentro de cada tipo de vista).

Al hacerlo, obtuvimos lo siguiente:

```python
Golfo: {'Rápida': 0.225, 'Media': 0.650, 'Dilatada': 0.125}
NoGolfo: {'Rápida': 0.278, 'Media': 0.500, 'Dilatada': 0.222}
```

📌 Fijate que **la suma de los valores en cada fila es igual a 1**.

Esto es así ya que la distribución de las proporciones está condicionada al tipo de vista (es decir, al total de la fila).

En cambio, si estuviéramos trabajando con frecuencias relativas sobre el total general, la suma de todos los valores de la tabla sería 1.

La tabla condicional nos permite **comparar las proporciones dentro de cada grupo** y analizar si la rapidez de venta varía según la vista.


📊 **¿Qué nos dice esta tabla?**

Si comparamos las proporciones de ventas rápidas, vemos que, aproximadamente:
- **23%** de las casas **con vista al golfo** se vendieron rápido.
- **28%** de las casas **sin vista al golfo** se vendieron rápido.

Esto sugiere que, en proporción, las casas sin vista al golfo se vendieron un poco más rápido.

Por eso, podemos decir que **no parece haber una asociación clara entre tener vista al golfo y vender más rápido**.

### 🟦 Gráfico de barras agrupadas

En la clase anterior vimos que, para resumir gráficamente un único conjunto de datos cualitativos, podíamos usar un gráfico de barras.

Del mismo modo, para comparar dos conjuntos de datos cualitativos a la vez, existe el **gráfico de barras agrupadas**.

Este gráfico tiene estas características:
- Cada **grupo** en el eje horizontal representa una categoría de uno de los conjuntos.
- Dentro de cada grupo, se muestran **barras** para las categorías del otro conjunto.

En el ejemplo que vimos graficamos la **rapidez de venta** según el **tipo de vista**.

En el eje horizontal del gráfico  las categorías de rapidez de venta: *Rápida*, *Media* y *Dilatada*.

Dentro de cada categoría, comparamos la proporción de casas según el tipo de vista: Vista al Golfo (*Golfo*) o Sin Vista al Golfo (*NoGolfo*).

Este gráfico muestra la **misma información** que la tabla de contingencia condicional, pero de forma visual, lo que facilita la comparación.

Recordá que los valores que graficamos son **proporciones condicionales**, es decir, se calculan **dividiendo cada celda por el total de su fila** en la tabla de contingencia (una fila por cada tipo de vista).

Como ya habíamos hecho ese cálculo, ahora simplemente lo tomamos del diccionario `tabla_contingencia_condicional`.

### 💻 Gráfico de barras agrupadas con Matplotlib

Para hacer este gráfico usamos la función `plt.bar` como antes, pero **la llamamos tres veces**, una para cada grupo:

```python
plt.bar(x_ado, uso_adolescentes, width=ancho, label="Adolescentes")
plt.bar(x_jov, uso_jovenes, width=ancho, label="Jóvenes")
plt.bar(x_adu, uso_adultos, width=ancho, label="Adultos")
```
Cada llamada grafica de las barras correspondientes al grupo etario ("Adolescentes", "Jóvenes" y "Adultos"):
- `x_ado`, `x_jov` y `x_adu` son listas que indican **dónde ubicar cada barra en el eje horizontal**. Las separamos un poco para que las barras de cada grupo no se superpongan.
- `uso_adolescentes`, `uso_jovenes` y `uso_adultos` contienen las **proporciones** de personas que eligieron cada red social, **dentro de cada grupo etario**.
- `width` define el **ancho de las barras**. En este ejemplo, usamos `ancho = 0.35` para que las barras queden bien visibles, una al lado de la otra.
- `label` nos permite darle un **nombre** a cada grupo para mostrarlo en la leyenda del gráfico.

Para mostrar qué color representa a cada grupo, usamos:
```python
plt.legend()
```

Esto agrega una **leyenda** que indica, por ejemplo, qué color corresponde a "Adolescentes", "Jóvenes" y "Adultos".

También, **personalizamos las etiquetas del eje horizontal** (eje X) del gráfico:
```python
plt.xticks(x, categorias_red)
```

En los argumentos, `x` es la lista de **posiciones numéricas** es una lista de posiciones numéricas sobre el eje X donde se colocarán las etiquetas. Por otro lado, `categorias_red` es la lista con los **nombres que queremos mostrar en esas posiciones**.
