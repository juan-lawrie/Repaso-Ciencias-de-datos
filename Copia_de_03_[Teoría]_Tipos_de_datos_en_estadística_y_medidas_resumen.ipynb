### 🔢 ¿Qué es la estadística?

La **estadística** es la herramienta que nos permite recopilar, organizar y analizar datos para entender mejor la realidad y tomar decisiones basadas en información.  

Muchas veces, al hablar de estadística, pensamos en cálculos como promedios o porcentajes. Sin embargo, su propósito va mucho más allá: nos ayuda a describir cómo se distribuyen los datos, analizar sus diferencias y, en muchos casos, sacar conclusiones sobre un grupo grande (población) usando solo una parte de él (muestra).  

En la actualidad, vivimos rodeados de datos. Encuestas, redes sociales, censos y registros digitales generan enormes cantidades de información que necesitan ser interpretadas correctamente. Por eso, el análisis estadístico es una herramienta clave en muchas áreas, como la ciencia, la economía, la salud y la tecnología.  

En esta clase, vamos a conocer los **tipos de datos en estadística**, cómo identificarlos según sus características y qué **medidas** disponemos para resumir la información de manera que tenemos. 🚀

**1. Datos cuantitativos (numéricos)**

Los datos cuantitativos miden o registran una cantidad numérica para cada persona u objeto y permiten realizar operaciones matemáticas, como sumas o promedios. Se dividen en dos tipos:

- **Cuantitativos discretos**: son aquellos que sólo pueden tomar valores enteros.
- **Cuantitativos continuos**: pueden tomar cualquier valor dentro de un rango, incluyendo decimales.

En la tabla:
- *Materias cursadas* es de tipo discreto pues cuenta la cantidad de materias cursadas, y se expresa en números enteros.
- *Rendimiento académico* es de tipo continuo. El rendimiento académico puede ser 8, 9 o 10, pero también puede incluir valores intermedios como 8.5, 9.1, etc.
- *Edad* es de tipo continuo. Aunque en la tabla venga dada con números enteros, 19, 20, si pensamos en su naturaleza (digamos tiempo) puede tomar cualquier valor dentro de un intervalo, por ejemplo 19.5 es 19 años y 5 meses (y podríamos seguir con semanas, días, minutos, segundos...).

💡 **Regla útil para datos cuantitativos**:

📌 Si es tiempo o medición → continua

📌 Si es conteo → discreta

⚠️ ¡Pero cuidado!

A veces las variables cuantitativas continuas se registran como valores enteros (en nuestro ejemplo, la *Edad* en años). Sin embargo, por su naturaleza siguen siendo variables cuantitativas continuas, ya que podrían medirse con mayor precisión.

**2. Datos cualitativos (categóricos)**

Los datos cualitativos agrupan a los casos en diferentes categorías, sin que exista una relación numérica entre ellas. Este tipo de datos se clasifica en dos tipos principales:

- **Cualitativos nominales**: son aquellos datos cuyas categorías no tienen un orden específico. Es decir, las categorías son solo etiquetas para identificar los casos, pero no se pueden organizar de manera jerárquica.
- **Cualitativos ordinales**: son datos cuyas categorías sí tienen un orden o jerarquía. Aunque los valores representan diferentes categorías, existe una relación de mayor a menor, o de mejor a peor, entre ellas.

Ejemplos en la tabla:
- *Usa redes sociales* y *Medio de transporte* son de tipo nominales. En estos casos, las categorías (por ejemplo, "Sí" o "No" para usar redes sociales, o "Auto", "Bicicleta", "Caminando" y "Colectivo" para el transporte) no tienen un orden específico, solo permiten agrupar los casos en diferentes grupos.
- *Satisfacción con la carrera* es de tipo ordinal. Las categorías de satisfacción (Baja, Media, Alta) tienen un orden jerárquico: "Baja" representa el nivel más bajo de satisfacción, seguido de "Media" y luego "Alta", lo que indica un grado de satisfacción creciente.

### 📊 Medidas para datos cualitativos

Cuando trabajamos con datos solemos manejar mucha información. Para poder interpretarla mejor, necesitamos herramientas que nos ayuden a resumir los datos en pocos números sin perder detalles importantes.

Las **medidas resumen** son números que nos permiten identificar rápidamente los aspectos más importantes de un conjunto de datos sin necesidad de tener que revisar cada observación una por una.

Para datos **cualitativos** categóricos, usamos tres medidas resumen principales:

- **Cantidad**: es el número de respuestas que pertenecen a cada categoría. Esto también se conoce como **frecuencia absoluta**. La suma de todas las frecuencias absolutas es igual al total de datos.  

- **Proporción**: se obtiene dividiendo la frecuencia absoluta de una categoría entre el total de datos. Es un número entre 0 y 1 que indica la fracción del total correspondiente a esa categoría. También se le llama **frecuencia relativa**, y la suma de todas las proporciones es 1.  

- **Porcentaje**: expresa la proporción en una escala de 0 a 100. Se obtiene multiplicando la proporción por 100. La suma de los porcentajes de todas las categorías es 100.  

📌 Un porcentaje del 25\% equivale a una proporción de 0.25. Para convertir un porcentaje en una proporción, lo dividimos entre 100. Para convertir una proporción en un porcentaje, la multiplicamos por 100.  

En el código que vimos arriba, la variable *Satisfacción con la carrera* tiene tres categorías: Baja, Media y Alta.

Sus frecuencias absolutas son:  

- Baja: 1  
- Media: 2  
- Alta: 2  

La suma de estas frecuencias es 1 + 2 + 2 = 5, que corresponde al total de datos disponibles.

Luego, calculamos la frecuencia relativa de cada categoría tanto en propociones como en porcentajes:

- Baja: 1/5 = 0.2 (20\%)
- Media: 2/5 = 0.4 (40\%)
- Alta: 2/5 = 0.4  (40\%)

Esta suma da 0.2 + 0.4 + 0.4 = 1 (o 100\% en porcentaje).

A partir de estos resultados, podemos interpretar que hay más personas satisfechas con la carrera (Media o Alta) que insatisfechas (Baja).

🤔 **¿Frecuencias absolutas o relativas?**

Calcular frecuencias nos ayuda a responder preguntas como:
- ¿Cuál es la categoría más y menos frecuente?
- ¿Cuántas personas/datos pertenecen a cada categoría?
- ¿Hay más personas en una categoría que en otra?
- ¿Qué proporción del total representa cada categoría?

Mientras que las **frecuencias absolutas** nos indican la cantidad exacta, las **frecuencias relativas** son más útiles para realizar comparaciones, especialmente cuando los conjuntos de datos tienen tamaños diferentes, algo que con las frecuencias absolutas es más complicado. Además, las frecuencias relativas son más fáciles de interpretar, ya que expresan los datos en proporciones o porcentajes.

En resumen, las **frecuencias absolutas** nos dicen **"cuántos"**, mientras que las **frecuencias relativas** nos muestran **"qué proporción"**, lo que las hace más útiles para análisis comparativos.

### 📊 Medidas para datos cuantitativos

Cuando trabajamos con datos **cuantitativos** (es decir, datos numéricos), disponemos de diversas **medidas resumen** para describir sus características más importantes. Estas medidas se dividen en dos tipos principales:

- **Medidas de tendencia central**, que indican cuál es el valor central del conjunto de datos.
- **Medidas de dispersión** o variabilidad, que describen qué tan dispersos o concentrados están los datos alrededor de ese valor central.

🎯 **Medidas de tendencia central**

Las medidas de tendencia central nos dan información acerca del **promedio** de un conjunto de datos, es decir, el valor característico o central del conjunto.

Hay diferentes formas para calcular el promedio, y la elección del método puede influir en la interpretación de los datos. En los medios de comunicación se suele hablar de "promedio" sin especificar qué tipo de cálculo se ha utilizado, lo que puede llevar a conclusiones erróneas.

Dos de las formar más comunes de obtener un valor central o promedio son: la **media** (que presentamos intuitivamente en la clase 1), y la **mediana**.

**1. Media**

La media se obtiene sumando todos los datos y luego dividiendo por la cantidad total de datos. Se representa con el símbolo $\bar{x}$ (se lee "equis barra") y la fórmula para un conjunto de datos, como $x_1, x_2, \ldots, x_n$, es

$$
\begin{aligned}
\bar{x} &= \frac{\text{Suma de los datos}}{\text{Número total de datos}} \\
\quad \\
\bar{x} &= \frac{x_1 + x_2 + \ldots + x_n}{n}
\end{aligned}
$$

La suma $x_1 + x_2 + \ldots + x_n$ se suele abreviar usando la letra griega **sigma mayúscula** $\sum$ (que significa **sumatoria**). En lugar de escribir vez de la suma de todos los datos, usamos $$\sum_{i=1}^n x_i$$ y lo leemos como: *la suma de todos los valores $x_i$ ("equis $i$"), con $i$ variando desde 1 hasta $n$*.

Así, la media de un conjunto de datos se puede escribir como:

$$ \bar{x} = \frac{\sum_{i=1}^n x_i}{n} \quad \text{o} \quad \bar{x} = \sum_{i=1}^n \frac{x_i}{n} $$


**2. Mediana**

La mediana es el punto central del conjunto de datos. Imaginá una línea
que divide algo en dos partes iguales, donde el 50% de los datos está por debajo de la mediana y el otro 50% está por encima de ella.

Para calcular la mediana seguimos estos pasos:
- Primero los **ordenamos los datos de menor a mayor**.
- Luego, **buscamos la posición central** que corresponde la mediana.

Si la cantidad de datos es par no hay una observación central, sino un par de
observaciones centrales. Por lo tanto, la mediana es el promedio de estos dos valores centrales.

La regla general para calcular la mediana de $n$ datos ordenados es:
- Si la cantidad de datos es **impar**, la mediana es el valor del centro, se encuentra en la posición $(n+1)/2$.
- Si la cantidad de datos es **par**, la mediana es el promedio de los dos valores centrales, se encuentran en las posiciones $n/2$ y $(n/2)+1$.

Para ejemplificar, si tenemos 5 datos, la mediana es el valor que está en la posición 3. Mientras que si tenemos 6 datos, la mediana es el promedio de los valores en las posiciones 3 y 4.

En el ejemplo que vimos con los datos de *Rendimiento académico* la media y la mediana coinciden en el valor 7.9, pero esto no es tan común:

$$
\begin{array}{c c c c c c}
    & & & \text{Media = Mediana} & & \\
    & \bullet & \bullet & \color{red} \bullet & \bullet & \bullet \\
    & 6.8 & 7.2 & \mathbf{7.9} & 8.5 & 9.1
\end{array}
$$

La diferencia clave entre ambas medidas es cómo utilizan los datos:
- La **media** usa todos los datos para el cálculo.
- La **mediana** usa sólo los valores centrales del conjunto.

Por esta razón, la mediana cambia menos que la media cuando hay **valores extremos** (muy altos o muy bajos en relación a la mayoría) en el conjunto. En estos casos, preferiremos la mediana por sobre la media para representar el valor central típico de la mayoría de los datos.


**¿Dónde se concentran la mayoría de los datos?**  

- Si la **mediana** está más a la **derecha** que la **media**, significa que la mayoría de los datos están **a la derecha** y hay algunos valores muy pequeños que están **tirando** la media hacia la izquierda.  
- Si la **mediana** está más a la **izquierda** de la **media**, significa que la mayoría de los datos están **a la izquierda** y algunos valores muy grandes están **empujando** la media hacia la derecha.

Cuando la **media** y la **mediana** son **similares**, indica que los datos están distribuidos de manera equilibrada, sin valores extremos que desajusten el centro de los datos.

🔀 **Medidas de dispersión**

Además de conocer el centro de los datos, también nos interesa saber **cuán dispersos** o **agrupados** están esos datos alrededor de su centro. Esto nos da una idea de cuánta **variabilidad** existe en el conjunto de datos.

Es normal que los datos tengan variabilidad, porque diferentes personas tienen diferentes características (peso, altura, etc.) y además porque diferentes mediciones de una misma característica pueden dar resultados ligeramente distintos debido a errores de medición.

Para analizar dispersión tenemos dos medidas muy conocidas: el **rango** y el **desvío estándar**.

**1. Rango**

El rango es una medida sencilla de la dispersión de los datos. Se calcula como la diferencia entre el valor máximo y el valor mínimo de los datos (máximo-mínimo).

$$
\begin{array}{c c c c c c}
    & \text{Mínimo} & & & & \text{Máximo} \\
    & \color{red} \bullet & \bullet & \bullet & \bullet & \color{red} \bullet \\
    & \mathbf{6.8} & 7.2 & 7.9 & 8.5 & \mathbf{9.1}
\end{array}
$$

Este valor nos indica la distancia entre el dato más grande y el más pequeño, dándonos una idea general de cuán dispersos están los datos.


**2. Desvío estándar**

El desvío estándar es una medida de dispersión basada en la media y utiliza todos
los datos. Representa la distancia promedio de cualquier punto del conjunto de datos a
su centro (medido por la media) y se denota con la letra $s$.

Para calcular el desvío promediamos la diferencia entre cada dato y la media,
elevadas al cuadrado. Como este resultado tiene las unidades al cuadrado, luego se saca
la raíz cuadrada.

Para un conjunto de $n$ datos:
1. Calculamos la **distancia de cada dato a la media**: $x_i - \bar{x}$.
2. **Elevamos al cuadrado** estas distancias: $(x_i - \bar{x})^2$.
3. Luego **promediamos** las distancias al cuadrado **dividiendo por $n$**. Este valor se llamda **varianza**

$$ s^2 = \frac{1}{n} \sum_{i=1}^n (x_i - \bar{x})^2$$

4. Por último, obtenemos el desvío calculando la **raíz cuadrada** de la varianza

$$
\begin{aligned}
s &= \sqrt{s^2} \\
\quad \\
s &= \sqrt{\frac{1}{n} \sum_{i=1}^n (x_i - \bar{x})^2}
\end{aligned}
$$

El desvío estándar nos indica, en promedio, cuánto se desvían los datos respecto a la media. Un valor mayor de $s$ significa que los datos están más dispersos.

$$
\begin{array}{c c c c c c}
    & & & \text{Media} & & \\
    & \color{orange} \bullet & \color{green} \bullet & \bullet & \color{green} \bullet & \color{orange} \bullet \\
    & 6.8 & 7.2 & \mathbf{7.9} & 8.5 & 9.1
\end{array}
$$

En los datos de *Rendimiento académico*, la media es 7.9. Los datos que están en **verde** están más cerca de la media, por lo tanto contribuyen menos al desvío estándar. En cambio, los datos en color **naranja** en naranja están más alejados, indicando mayor variabilidad. A mayor cantidad de valores alejados de la media, mayor será el desvío estándar.

Al igual que con la media y la mediana, también existen diferencias importantes entre el rango y el desvío estándar. En el ejemplo con los datos de *Rendimiento académico*, obtuvimos un rango de 2.7 y desvío estándar de 0.93, aproximadamente. Es muy difícil que estas medidas coincidan, más aún que para el caso entre la media y la mediana.

La diferencia clave entre ambas medidas es cómo utilizan los datos:
- El **rango** se calcula utilizando sólo dos valores del conjunto, el máximo y el mínimo.
- El **desvío estándar** considera todos los datos y también la media.

Dado que el rango sólo depende del valor máximo y mínimo, puede cambiar mucho si son **valores extremos**. Además, como utiliza sólo dos datos, no puede distinguir dos conjuntos con máximos
y mínimos coincidentes, pero uno tendrá la mayoría de sus valores mucho más concentrados que el otro. Por su parte, el desvío tiene en cuenta los datos, permitiendo así diferenciar conjuntos según qué tan dispersos están los valores alrededor de la media. Sin embargo, también puede verse afectado por valores extremos, aunque en menor medida que el rango.

El primer paso para calcular la mediana es ordenar los datos. Pero, dado que aún no sabemos ordenar un conjunto de datos automáticamente (lo aprenderemos más adelante), por ahora debemos hacerlo manualmente.

Para calcular la mediana con los datos de *Edad* seguimos estos pasos:

Para calcular estas medidas también nos vemos limitados por las herramientas que disponemos, pero esto igual nos sirve para comprender cada paso del proceso de cálculo, como nos pasó con la mediana.

Como aún no sabemos encontrar el máximo y mínimo valor en un conjunto, tenemos que asignarlos a mano. En el ejemplo con los datos de *Edad* asignamos manualmente el máximo y mínimo valor a las variables `max_edad` y `min_edad`, respectivamente. Haciendo la diferencia entre estos valores, obtuvimos el **rango**.

El caso del desvío estándar implica aplicar las operaciones dato por dato y luego combinar los resultados:

1. El primer requisito es obtener la **media**, como ya la calculamos antes, la tenemos disponible en la variable `media_edad`.
2. Luego, para cada dato calculamos su **diferencia con la media**, resultado que guardamos en las variables `dist1_media`, `dist2_media`, etc.
3. Luego tuvimos que elevar cada diferencia al **cuadrado** y guardamos estos valores en las variables `dist1_media_2`, `dist2_media_2`, etc. (el "2" final en el identificador indica "cuadrado").
4. Después, sumamos los cuadrados y dividimos por el total de datos `n`. Esto nos da la **varianza**.
5. Por último, aplicamos la raíz cuadrada para obtener el **desvío estándar**, elevando la varianza a la 1/2 o 0.5.
