## 🛍️ Programa de beneficios 1.0

Una tienda local ha implementado un **programa de beneficios** basado en puntos para premiar a su clientela más fiel.

El programa de beneficios funciona de la siguiente manera:
- **Ganás 10 puntos por cada \$5.000 de compra**.
- **Si sos cliente habitual, recibís 5 puntos extra**.
- **Si alcanzás 100 puntos o más, obtenés un 20\% de descuento en tu compra**.

📌 Como caso de estudio, contamos con la siguiente información de compra de una persona:

- Categoría de cliente: Habitual
- Puntos antes de la compra: 70
- Monto de compra: \$18.500

¿Como podemos hacer para que la compu verifique si corresponde un descuento y lo calcule? 🤔

Analizá el siguiente código y tratá de entender qué hace cada linea:

````python
categoria_cliente = "Habitual"
puntos_actuales = 70
monto_compra = 18500

puntos_ganados = (monto_compra // 5000) * 10
puntos_totales = puntos_actuales + puntos_ganados

if categoria_cliente == "Habitual":
    puntos_totales = puntos_totales + 5

if puntos_totales >= 100:
    monto_descuento = monto_compra * (20 / 100)
else:
    monto_descuento = 0

monto_a_pagar = monto_compra - monto_descuento

print("Puntos totales:", puntos_totales)
print("Descuento aplicado:", monto_descuento)
print("Monto a pagar:", monto_a_pagar)
````

### 🔣 Operadores relacionales

En el código que vimos usamos unos operadores especiales como `==` y `>=` para comparar valores. Estos operadores se conocen como **operadores relacionales**.

Por ejemplo, cuando escribimos `es_habitual == "Habitual"`, usamos el operador "igual a" `==` para decirle a la compu que verifique si el valor de la variable `es_habitual` es **igual a** `"Habitual"`.

En otro ejemplo, usamos el operador "mayor o igual que" `>=` en la expresión `puntos_totales >= 100` para indicarle a la compu que verifique si el valor de `puntos_totales` es **mayor o igual** a 100.

En Python tenemos seis operadores relacionales principales:

|Operador|Descripción|Ejemplo|Interpretación|
|:--|:--|:--|:--|
|`==`|Igual a|`puntos_totales == 100`|¿Tiene exactamente 100 puntos?|
|`!=`|Distinto de|`puntos_totales != 100`|¿No tiene exactamente 100 puntos?|
|`>`|Mayor que|`puntos_totales > 100`|¿Tiene más de 100 puntos?|
|`<`|Menor que|`puntos_totales < 100`|¿Tiene menos de 100 puntos?|
|`>=`|Mayor o igual que|`puntos_totales >= 100`|¿Tiene 100 o más puntos?|
|`<=`|Menor o igual que|`puntos_totales <= 100`|¿Tiene 100 o menos puntos?|

Cuando comparamos valores obtenemos como resultado un valor **booleano**, es decir, `True` o `False` (verdadero o falso). A este tipo de expresiones que sólo pueden evaluarse como verdadero o falso se les denomina **expresiones lógicas**.

### 🔀 Ramificación

En la clase anterior hablamos de programas que fluyen secuencialmente, es decir, que ejecutan las instrucciones (o sentencias) de forma lineal una tras otra, sin saltar ni repetir ninguna.

Pero ahora, en el código del programa de beneficios, hay algo más interesante: la **ramificación** (o *branching*, en inglés). Esto en programación se refiere a la capacidad de un programa para **tomar decisiones** y ejecutar diferentes bloques de código según ciertas condiciones. De esta forma, el flujo de ejecución deja de ser lineal y se divide en diferentes caminos dependiendo de valores o condiciones específicas.

### 🔀 Estructuras de control condicionales

El cambio en el flujo de ejecución del programa se logra mediante **estructuras de control**. En el ejemplo usamos `if` y `if-else`, que son estructuras de control **condicionales**. Estas estructuras nos permiten introducir la **ramificación** en el programa, lo que hace que el flujo de ejecución se divida y siga diferentes caminos dependiendo de si se cumple o no una **condición**.

Una condición es una expresión lógica (como las comparaciones que vimos antes) que se evalúa y devuelve un resultado **booleano**. Decimos que la condición "se cumple" cuando el resultado de la evaluación es `True` (verdadero) y que "no se cumple" cuando el resulado es `False` (falso).



Por ejemplo, cuando escribimos:
```python
if categoria_cliente == "Habitual":
    puntos_totales = puntos_totales + 5
```

Le estamos diciendo a la compu que verifique la condición `categoria_cliente == "Habitual"`. Si el resultado es `True`, entonces la compu ejecutará la línea `puntos_totales = puntos_totales + 5` que está dentro del `if`. Pero si es `False`, la compu omitirá esta línea y continuará con el resto del código.

A veces queremos ofrecer una **alternativa** en caso de que la condición no se cumpla. Este es el caso de cuando escribimos:
```python
if puntos_totales >= 100:
    monto_descuento = monto_compra * (20 / 100)
else:
    monto_descuento = 0
```

Acá le indicamos a la compu que verifique la condición `puntos_totales >= 100`. Si es `True`, la compu ejecutará la línea `monto_descuento = monto_compra * (20 / 100)` dentro del `if`. Pero si es `False`, la compu ejecutará la línea `monto_descuento = 0` que está dentro del `else`. Luego sí continuará con el resto del código.

Como podrás ver, en una estructura de control condicional el programa elegirá qué bloque de código ejecutar dependiendo de si la condición definida se cumple o no:

- El bloque `if` evalúa una condición y ejecuta el código dentro de él solo si la condición es **verdadera**.
- El bloque `else` se ejecuta cuando la condición del `if` es **falsa**.
- Si no hay un `else`, el código dentro del `if` se **omite** (no se ejecuta).

### 🐞  Errores comunes en operadores

En el ejemplo vemos que la compu no acepta operadores mal escritos y nos lo indica con **mensajes de error**.

En el primer caso, para comparar si dos valores son iguales, el operador correcto es `==`, no `===`. Cuando escribimos `categoria_cliente === "Habitual"`, la compu nos marca un error de **sintaxis** porque `===` no es un operador válido. Lo correcto es escribir `categoria_cliente == "Habitual"`.

Otro error común lo vemos cuando queremos saber si un valor es mayor o igual que otro. El operador correcto es `>=` pero si lo escribimos al revés como en `puntos_totales => 100`, la compu nos dirá que la sintaxis no es válida. La forma correcta es `puntos_totales >= 100`.

Por otro lado, en la primera clase vimos que existen errores que **no generan un mensaje de error**, pero hacen que el programa no funcione como queremos. A esto le llamamos **bug**.

Por ejemplo, si queremos verificar si el descuento supera el tope máximo y escribimos `monto_descuento != 5000`, vemos que la expresión esta bien sintácticamente pero **no representa lo que realmente queremos verificar**. Como buscamos comprobar si el descuento es **mayor** al límite, el operador correcto es `>`. La expresión correcta sería `monto_descuento > 5000`.



### ➡️ Indentación en condicionales

¿Te diste cuenta que el código qué está dentro de los condicionales `if` y `else` tiene espacios a la izquierda?

Si recordás, en la primera clase aprendimos que esto se llama **indentación**, y dijimos que en Python es un aspecto esencial para estructurar el código de manera correcta. Bueno, acá vemos por qué es tan importante.

Cuando usamos estructuras de control como `if` o `else` hay **dos reglas** clave:
1. Colocar **dos puntos** (`:`) al final del `if` y `else`.
2. **Indentar** el código que pertenece a la estructura para indicarle a la compu que líneas forman parte del bloque y se deben ejecutar luego de evaluar la condición.



Por ejemplo, en este código:

```python
if categoria_cliente == "Habitual"
puntos_totales = puntos_totales + 5
```

Faltan los dos puntos al final del `if` y la línea dentro del `if` no está indentada. Lo correcto sería así:


```python
if categoria_cliente == "Habitual":
    puntos_totales = puntos_totales + 5
```

Otro ejemplo es este código:

```python
if puntos_totales >= 100
monto_descuento = monto_compra * (20 / 100)
else
monto_descuento = 0
```

Acá faltan los dos puntos al final de la línea del `if` y del `else` y la indentación dentro de cada bloque. Luego de corregirlo quedaría así:

```python
if puntos_totales >= 100:
    monto_descuento = monto_compra * (20 / 100)
else:
    monto_descuento = 0
```

📌 Cuando no respetamos las reglas de indentación la compu nos mostrará un mensaje de error. Lo más común es indentar usando **cuatro espacios** o un **tabulador** (tecla Tab en el teclado).

### 🔤 Comparación con cadenas de texto



Las cadenas de texto (*strings*) se pueden comparar utilizando los operadores relacionales. Por ejemplo, en `categoria_cliente == "Habitual"` usamos el operador igual `==`.

La comparación se realiza caracter por caracter siguiendo el **orden lexicográfico**, que está basado en la tabla **ASCII** (o **Unicode** para caracteres especiales). En esta tabla, cada caracter tiene un número asociado. Por ejemplo, las letras mayúsculas (A-Z) tienen valores de 65 a 90, las letras minúsculas (a-z) tienen valores de 97 a 122 y los números (0-9) van de 48 a 57.

Al comparar dos cadenas, la compu analiza los caracteres **uno a uno**, **de izquierda a derecha**, usando sus valores numéricos ASCII para determinar el orden. Si una cadena es prefijo de otra, la más corta se considera menor.

Ahora podemos ver que las mayúsculas y minúsculas **no son iguales** porque tienen valores ASCII distintos. En el fondo, la comparación de strings es simplemente una comparación de números.

## 🛍️ Programa de beneficios 3.0

Después de analizar el rendimiento del programa durante varios meses, la tienda ha decidido implementar una nueva actualización.

A partir de ahora, **los clientes no ganarán puntos por sus compras, sino por sus visitas**.

Las nuevas reglas para los **puntos** por visitas anuales son las siguientes:
- **Más de 10 visitas: 20 puntos**.
- **Entre 5 y 10 visitas: 5 puntos**.
- **Menos de 5 visitas: 1 punto**.


Los **descuentos** se aplicarán de la siguiente manera:
- **100 puntos o más: 20\% de descuento**.
- **Si además el gasto anual es de \$150.000 o más: 10\% de descuento adicional**.
- **El descuento del 10\% sólo es aplicable si corresponde el descuento del 20\%**.


📌 Tenemos la siguiente información sobre una compra:

- Visitas en el último año: 15
- Gasto en compras en el último año: \$120.000
- Puntos antes de la compra: 85
- Monto de compra: $11.000


¿Cómo podemos hacer para que la compu calcule los puntos por visitas y aplique el descuento correspondiente? 🤔

Mirá el siguiente código y **pensá** qué hace cada linea. Luego ejecutá la celda y analizá los resultados:

````python
visitas_ultimo_anio = 15
gasto_ultimo_anio = 120_000
puntos_iniciales = 85
monto_compra = 11_000

if visitas_ultimo_anio > 10:
    puntos_por_visitas = 20
elif visitas_ultimo_anio >= 5:
    puntos_por_visitas = 5
else:
    puntos_por_visitas = 1

puntos_totales = puntos_iniciales + puntos_por_visitas

if puntos_totales >= 100:
    descuento = 0.20
    if gasto_ultimo_anio >= 150_000:
        descuento = descuento + 0.10
else:
    descuento = 0

monto_descuento = monto_compra * descuento
monto_a_pagar = monto_compra - monto_descuento

print("Puntos totales:", puntos_totales)
print("Descuento aplicado:", monto_descuento)
print("Monto a pagar:", monto_a_pagar)
````

### 🔢 Notación con guion bajo para números grandes

En el código de arriba vimos una forma alternativa de escribir números utilizando el **guión bajo**. Por ejemplo, en lugar de `120000`, escribimos `120_000`.

Vimos que esta notación **no afecta el valor** del número porque la compu ignora los guiones bajos al interpretar el número. Sin embargo, es una manera de representar el mismo número de manera más legible.

Esto es muy útil especialemente en **números grandes**. Por ejemplo, `1_000_000_000` es mucho más claro que `1000000000`.

### 🔀 Condicionales con elif

Cuando en una decisión tenemos **más de dos alternativas**, podemos usar la estructura `if-elif-else` para crear más ramificaciones y permitir que el programa elija entre diferentes bloques de código según las condiciones que se vayan cumpliendo.

Además del `if` y `else` que ya conocemos, podemos incluir **uno o más** `elif` para manerar distintos escenarios intermedios.

Por ejemplo, en el cálculo de los puntos por visitas asignamos diferentes valores según la cantidad de visitas en el último año:

```python
if visitas_ultimo_anio > 10:
    puntos_por_visitas = 20
elif visitas_ultimo_anio >= 5:
    puntos_por_visitas = 5
else:
    puntos_por_visitas = 1
```

En este caso, la compu primero evalúa la primera condición del `if`. Si es `True`, ejecuta su bloque de código y **omite** las demás condiciones. Si es `False`, pasa a evaluar el `elif`.
Si ninguna de las condiciones anteriores se cumple, ejecuta el código dentro del `else` como última opción.

### 🔗 Condicionales anidados

En programación podemos hacer muchas cosas con los condicionales, una de ellas es incluir un `if` dentro de otro `if`. A esto se le llama **anidamiento** y es útil cuando una decisión depende de que otra se cumpla primero.

En el código que vimos, el descuento adicional sólo se calcula cuando el cliente acumuló 100 o más puntos:

```python
if puntos_totales >= 100:
    descuento = 0.20
    if gasto_ultimo_anio >= 150_000:
        descuento = descuento + 0.10
else:
    descuento = 0
```

Primero, la compu evalúa si el cliente tiene 100 o más puntos. Si es `True`, aplica un 20% de descuento y revisa si el gasto anual es 150.000 o más. Si también se cumple esta segunda condición, suma un 10% adicional. Sin embargo, si la primera condición no se cumple, el descuento será 0.

Notemos que cada `if` anidado debe estar **indentado un nivel más** que el anterior para que la compu lo interprete correctamente.

## 🛍️ Programa de beneficios 4.0

La tienda ha decidido mejorar aún más su programa de beneficios, agregando nuevas condiciones para otorgar descuentos especiales.

Ahora, además de los puntos acumulados y el monto gastado en el último año, también se considerará si la persona realizó compras durante eventos especiales o si es un cliente VIP.

Los **puntos** por visitas se otorgan igual que antes:
- **Más de 10 visitas: 20 puntos**.
- **Entre 5 y 10 visitas: 5 puntos**.
- **Menos de 5 visitas: 1 punto**.

Los nuevos **descuentos** se aplican descuentos de la siguiente manera:
- **15\% de descuento si acumula 100 puntos o más y el gasto anual es de \$150.000 o más**.
- **10\% de descuento extra si la persona compró durante eventos especiales o es cliente VIP**.
- **No se aplica ningún descuento si la compra no supera los \$10.000**.

📌 Información de compra de una persona:

- Visitas en el último año: 8
- Gasto en compras en el último año: \$95.000
- Puntos antes de la compra: 70
- Monto de compra: \$15.000
- Cliente VIP: Sí
- Compró durante eventos especiales: No

¿Cómo podemos hacer para que la compu determine el total de puntos y el descuento aplicable para este caso? 🤔


````python
visitas_ultimo_anio = 8
gasto_ultimo_anio = 95_000
puntos_iniciales = 70
monto_compra = 15_000
es_vip = True
compro_en_evento = False

if visitas_ultimo_anio > 10:
    puntos_por_visitas = 20
elif visitas_ultimo_anio >= 5:
    puntos_por_visitas = 5
else:
    puntos_por_visitas = 1

puntos_totales = puntos_iniciales + puntos_por_visitas

descuento = 0
if puntos_totales >= 100 and gasto_ultimo_anio >= 150_000:
    descuento = descuento + 0.15

if compro_en_evento or es_vip:
    descuento = descuento + 0.10

if not (monto_compra > 10_000):
    descuento = 0

monto_descuento = monto_compra * descuento
monto_a_pagar = monto_compra - monto_descuento

print("Puntos totales:", puntos_totales)
print("Descuento aplicado:", monto_descuento)
print("Monto a pagar:", monto_a_pagar)
````

### 🔣 Operadores lógicos

En el código anterior vemos que podemos combinar varias expresiones lógicas en una sola usando los operadores `and`, `or` y `not`. Estos operadores se llaman **operadores lógicos** y permiten evaluar condiciones y tomar decisiones en función de su resultado.

Los operadores lógicos funcionan de la siguiente manera:
- `and`: evalúa si **ambas** condiciones son verdaderas al mismo tiempo.
- `or`: evalúa si **al menos una** de las condiciones es verdadera.
- `not`: **niega** una condición, es decir, cambia una condición verdadera a falsa y viceversa.

Ejecutando las celdas que siguen podés probar el comportamiento de cada uno de estos operadores:

````python
print(True and True)   # True
print(True and False)  # False
print(False and True)  # False
print(False and False) # False
````

````python
print(True or True)   # True
print(True or False)  # True
print(False or True)  # True
print(False or False) # False
````

````python
print(not True)  # False
print(not False) # True
````

Como podemos ver en los resultados, `and` devuelve `True` sólo si ambas condiciones son `True`. Por su parte, `or` devuelve `True` si al menos una de las condiciones es `True`. Por último, `not` devuelve `True` si la condición es `False`, y `False` si la condición es `True`.

En nuestro programa de beneficios usamos `and` en la expresión `puntos_totales >= 100 and gasto_ultimo_anio >= 150_000`. En este caso, el resultado será `True` sólo si **ambas** condiciones son `True`.

Por otro lado, usamos `or` en la expresión `compro_en_evento or es_vip`. Acá, el resultado será `True` si **al menos una** de las condiciones es `True`.

Por último, utilizamos `not` en la expresión `not (monto_compra > 10_000)`. El resultado será `True` si la condición es `False`.

📚 **Un poco de teoría**

Las expresiones lógicas, como hemos visto, devuelven un valor booleano (`True` o `False`). El término "booleano" hace referencia a **George Boole**, un matemático del siglo XIX que sentó las bases del **álgebra booleana**. Este sistema lógico es fundamental en la programación y en el diseño de circuitos digitales, ya que permite operar con i**nformación binaria**.

En este contexto, la información se representa mediante dos estados posibles: **`1` (verdadero) o `0` (falso)**. Los operadores lógicos permiten combinar estos valores para construir expresiones más complejas:

- **`and` (Y o "conjunción")**:  es una función de dos entradas que realiza una **multiplicación lógica**. Devuelve `1` (verdadero) únicamente cuando **ambas** entradas son `1`. En todos los demás casos, el resultado es `0` (falso).
- **`or` (O o "disyunción")**: también es una función de dos entradas, pero actúa como una **suma lógica**. Devuelve `1` si **al menos una** de las entradas es `1`. Solo devuelve `0` cuando **ambas** entradas son `0`.
- **`not` (No o "negación")**: es una función de una sola entrada que **invierte** el valor de la entrada. Si la entrada es `1`, devuelve `0`, y si la entrada es `0`, devuelve `1`.

---

🔣 **Precedencia de operadores lógicos**

Los operadores lógicos `and`, `or` y `not` se pueden combinar para formar expresiones más complejas. La compu evalúa estas expresiones siguiendo las reglas de **precedencia**:

1. `not`
2. `and`
3. `or`

Si una expresión contiene **paréntesis**, estos tendrán la mayor prioridad y se evaluará primero lo que está entre paréntesis. Además, si hay operadores repetidos con la misma precedencia, la evaluación se realizará de **izquierda a derecha**.


Mirá estas expresiones lógicas y **pensá** que resultado deberían dar:

````python
not True and False or True
````

````python
not (True and False or True)
````

🔣 **Precedencia de operadores en expresiones mixtas**

Cuando en una expresión combinamos operadores aritméticos, relacionales y lógicos la compu sigue un orden de evaluación basado en la siguiente regla de precedencia:

1. Operadores **aritméticos**: `**`, `*`, `/`, `//`,`%`, `+`, `-`
2. Operadores **relacionales**: `<`, `<=`, `>`, `>=`, `==`, `!=`
3. Operadores **lógicos**: `not`, `and`, `or`



Mirá estas expresiones lógicas y pensá que resultado deberían dar:

````python
10 + 5 > 12 and 8 % 3 == 2 or not False
````

````python
not (10 + 5 > 12 and 8 % 3 == 2) or False
````
