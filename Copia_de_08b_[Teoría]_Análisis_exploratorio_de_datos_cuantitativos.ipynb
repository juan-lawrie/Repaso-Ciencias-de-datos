## 📱 Tiempo de uso de redes sociales

Suppongamos que se encuestó a 30 personas y se les preguntó cuántos minutos al día usas redes sociales.

Los datos son los siguientes:

````python
tiempos_redes = [
    65, 80, 60, 75, 50, 90, 30, 120, 110, 65,
    70, 85, 95, 55, 40, 100, 55, 20, 80, 105,
    60, 80, 75, 95, 100, 75, 45, 70, 85, 95
]
````

Usá las funciones debajo para calcular la media y el desvío estándar de los datos:

````python
def calcular_media(lista):
    return sum(lista) / len(lista)

def calcular_desvio(lista):
    media = calcular_media(lista)

    suma_diferencias = 0
    for x in lista:
        dif = x - media
        suma_diferencias += dif ** 2

    desvio = (suma_diferencias / len(lista)) ** (1/2)
    return desvio
````

````python
media_tiempos = calcular_media(tiempos_redes)
desvio_tiempos = calcular_desvio(tiempos_redes)

print("Media de tiempos:", media_tiempos)
print("Desvío estándar de tiempos:", desvio_tiempos)
````

### 📐 Medidas de tendencia central y dispersión

Recordemos que para **datos cuantitativos** tenemos dos tipos de medidas para resumirlos.

📊 **Medidas de tendencia central**

Nos indican **dónde se concentran los datos**, es decir, cuál es el "valor típico" o "valor central".

- **Media**: es el **promedio** y se calcula sumando todos los valores y dividiendo por la cantidad total.

 👉 Muy **sensible** a valores extremos (atípicos o *outliers*).

- **Mediana**: es el valor que se encuentra en el **centro** cuando los datos están **ordenados**.

 👉 Es más **robusta** frente a valores extremos (atípicos o *outliers*).

📈 **Medidas de dispersión o variabilidad**

Nos dicen **qué tan dispersos** están los datos, si están muy cerca o lejos del centro.

- **Rango**: es la **diferencia** entre el **valor más alto** y el **más bajo**.

 👉 Es muy fácil de calcular, pero depende mucho de los extremos.

- **Desvío estándar**: mide, en promedio, **cuánto se alejan** los datos de la **media**.

 👉 Un valor alto indica que los datos están muy dispersos.

- **Cuartiles y rango intercuartílico (IQR)**: dividen los datos **ordenados** en **cuatro partes iguales** (Q1, Q2 y Q3).

 El IQR (Q3 − Q1) muestra la **dispersión** de la “zona media” de los datos (el 50% central) alrededor de la **mediana**.

 👉 Es muy útil para **detectar valores atípicos** y analizar la distribución sin que se vea afectado por valores extremos.


📌 Recordá que el cuadrado del desvío estándar nos da la **varianza**.

### 📊 Histograma

El **histograma** es uno de los gráficos más usados para analizar **datos cuantitativos**.

A diferencia de los gráficos de barras, no muestra categorías sino **rangos de valores numéricos** (intervalos).

Los datos numéricos se agrupan en **intervalos** iguales, llamados *bins* (por ejemplo: de 20 a 30, de 30 a 40, etc.). El **ancho de cada barra** depende del número de bins que se elijan: menos bins da lugar a barras más anchas, mientras que más bins producen barras más finas.

La **altura de cada barra** muestra la **densidad relativa**, es decir, la proporción de datos por unidad del intervalo.

📐 **¿Por qué importa el área de la barra?**

En un histograma lo que **representa la proporción de datos** (frecuencia relativa) no es la altura, sino **el área completa del rectángulo** que forma una barra.

Por lo tanto, para obtener la **frecuencia relativa** (el aŕea del rectángulo) hay que **multiplicar el ancho por la altura (densidad) de la barra**.

👉 **Lo importante es el área, no solo la altura**.

Por eso, dos barras pueden tener distinta altura, pero si su base es más ancha, pueden representar la misma cantidad de datos.

📈 **¿Para qué sirve un histograma?**

El histograma es muy útil para:
- Ver **dónde se concentran y cuán dispersos están los datos**.
- Observar la **forma general de la distribución**.
- Detectar posibles **valores extremos** o atípicos.

En cuanto a la **forma** que tienen los datos nos referimos a si el histogrma es **simétrico** o está **inclinado** hacia un lado.

Algunas formas comunes son:
- **Simétrica y acampanada**: los datos están **distribuidos de manera equilibrada** alrededor del centro y tienen forma de "campana".
- **Asimétrica a la derecha**: hay **muchos datos con valores bajos**, y pocos valores altos que "empujan" la cola hacia la derecha.
- **Asimétrica a la izquierda**: hay **muchos datos con valores altos**, y pocos valores bajos que "empujan" la cola hacia la izquierda.

### 💻 Histograma con Matplotlib

Para hacer el histograma usamos la función `plt.hist` de Matplotlib:

```python
plt.hist(tiempos_redes, bins=10, density=true, edgecolor="black")
```

Esta función tiene varios argumentos pero por ahora sólo nos interesan algunos:
- `tiempos_redes`: es la lista con los valores que queremos representar.
- `bins`: define cómo se agrupan los datos en intervalos en el eje horizontal.
- `density`: hace que el histograma muestre densidades relativas o cantidades absolutas.
- `edgecolor` (opcional): agrega bordes a las barras para verlas mejor.


💡 El parámetro `bins` permite especificar:
- un **número** (por ejemplo, `bins=10`) que indica **cuántos intervalos se usarán**, o
- una **lista con los límites exactos de los intervalos** (por ejemplo, `bins=[20, 40, 60, 80, 100, 120]`).

📏 **Escala de densidad en el histograma**

El agumento `density` de la función `plt.hist` modifica **la escala del eje vertical** del histograma:
- `density=False` (valor por defecto): muestra la **frecuencia absoluta**, es decir, la cantidad de datos que caen en cada intervalo (o *bin*).
- `density=True`: muestra la **densidad relativa**, es decir, la proporción de datos **por unidad de ancho de intervalo**.

💡 Cuando `density=True`, el área total bajo todas las barras del histograma es **igual a 1**, lo que permite interpretar las alturas como densidades. Además:
- Nos permite **comparar distribuciones** con distintos tamaños de muestra.
- Hace visible la **forma de la distribución** sin que dependa del número total de datos.

🧮 **¿Cómo se interpreta la densidad?**

La **densidad** en un histograma no se interpreta como una frecuencia directa, sino como una **medida de concentración de los datos por unidad de longitud (en nuestro caso la medida de longitud es el ancho de barra)**.

La altura de la barra no nos dice cuántos o qué proporción de datos hay en esa barra, sino que nos dice qué tan concentrados están los datos en esa barra.

## 🏢 Edificios

Con el objetivo de analizar las características de los edificios en su cartera, una empresa inmobiliaria estudió la cantidad de pisos que poseen 50 de sus edificios seleccionados al azar.

Los datos recolectados (cantidad de pisos por edificio) son los siguientes:

````python
pisos = [
    3, 4, 5, 6, 3, 7, 8, 4, 5, 6,
    9, 10, 4, 5, 6, 4, 12, 3, 7, 6,
    4, 8, 5, 5, 4, 6, 9, 15, 20, 25,
    4, 5, 3, 6, 7, 4, 5, 3, 8, 5,
    6, 4, 4, 5, 7, 6, 9, 11, 13, 6
]
````

Usá las funciones debajo para calcular el valor mínimo, máximo y los cuartiles de los datos:

````python
def calcular_mediana(lista):
    lista_ordenada = lista.copy()
    lista_ordenada.sort()

    n = len(lista_ordenada)

    if n % 2 == 0:
        mediana = (lista_ordenada[n // 2 - 1] + lista_ordenada[n // 2]) / 2
    else:
        mediana = lista_ordenada[n // 2]

    return mediana

def calcular_cuartiles(lista):
    datos = lista.copy()
    datos.sort()

    n = len(datos)
    if n % 2 == 0:
        mitad_inferior = datos[:n // 2]
        mitad_superior = datos[n // 2:]
    else:
        mitad_inferior = datos[:n // 2]
        mitad_superior = datos[n // 2 + 1:]

    Q1 = calcular_mediana(mitad_inferior)
    Q2 = calcular_mediana(datos)
    Q3 = calcular_mediana(mitad_superior)

    return Q1, Q2, Q3
````

````python
minimo = min(pisos)
maximo = max(pisos)
q1, q2, q3 = calcular_cuartiles(pisos)

print("Mínimo:", minimo)
print("Q1:", maximo)
print("Q2:", q1)
print("Q3:", q2)
print("Máximo:", q3)
````

### 📦 Diagrama de caja

Un **diagrama de caja** (o *boxplot*, en inglés) es una forma gráfica de resumir un conjunto de **datos cuantitativos**.

Permite ver de forma rápida:
- La **dispersión** (qué tan separados están los datos).
- La **asimetría** (forma de la distribución).
- La posible presencia de **valores atípicos** (*outliers*)

También es conocido como **diagrama de caja y bigotes**, porque tiene **una caja** (*box*) en el medio y **dos bigotes** (*whiskers*) a los lados.

✂️ **¿Qué muestra el boxplot?**

Un boxplot se basa en los **cinco números resumen** de un conjunto de datos:
- **Mínimo** (*m*): el valor más pequeño no atípico.
- **Q₁ (primer cuartil)**:  25% de los datos están por debajo.
- **Q₂ (mediana)**: 50% de los datos están debajo.
- **Q₃ (tercer cuartil)**: 75% de los datos están por debajo.
- **Máximo** (*M*): valor más grande no atípico.

📦 La **caja** va desde los cuartiles Q1 hasta Q3 y dentro tiene una línea que marca la **mediana**.

📏 Los **bigotes** se extienden desde los bordes de la caja hasta los valores máximo y mínimo, mientras no sean valores **atípicos** (es decir, se encuentren dentro de `1.5 * IQR`).

<div style="display: flex; justify-content: space-between;"> <img src="https://drive.google.com/uc?id=1ih62Esd4CpMnXWfPuYXnwyjYfsb1iaRU" width="700" />





✏️ **Construcción del boxplot**

📦 **La caja**:

- Se calcula el **mínimo** $m$ y el **máximo** $M$ de los datos.
- Se dibuja una **caja** desde $Q_1$ hasta $Q_3$.
- Dentro de la caja se marca la **mediana** con una línea.
- La **altura de la caja** (en vertical) o su **ancho** (en horizontal) representa el **rango intercuartil**:

$$ IQR = Q3 - Q1$$


📏 **Los bigotes**:

Se definen los **límites teóricos** para considerar un valor como no atípico según la regla del IQR:
$$L_{\text{inf}} = Q_1 - 1.5 \cdot IQR$$
$$L_{\text{sup}} = Q_3 + 1.5 \cdot IQR$$

Los **extremos reales de los bigotes** se calculan como:

- Si $m \ge L_{\text{inf}}$, entonces el bigote se corta en $m$. En cambio si $m < L_{\text{inf}}$, entonces el bigote se corta en el menor elemento del conjunto de datos que es mayor o igual que $L_{\text{inf}}$.

- Si $M\leq L_{\text{sup}}$, entonces  el bigote se corta en $M$. En cambio si $M>L_{\text{sup}}$, entonces el bigote se corta en el mayor elemento del conjunto de datos que es menor o igual que $L_{\text{sup}}$.

Es decir, los bigotes marcan el mínimo y el máximo de los datos, excepto cuando están muy alejados de la caja.


🔴 **Valores atípicos**:
- Todo valor **fuera de esos límites teóricos** se considera **atípico** y se marca con un punto o asterisco fuera de los bigotes.





👀 **¿Qué se puede observar con un boxplot?**

En cuanto a la **forma de la distribución** (simetría/asimetría), podemos decir:
- **Simétrica y acampanada**: la mediana está en el **centro de la caja** y los **bigotes son similares** en longitud.
- **Asimétrica a la derecha**: la mediana está **más cerca de Q1** y el **bigote derecho es más largo**.
- **Asimétrica a la izquierda**: la mediana está **más cerca de Q3** y el **bigote izquierdo es más largo**.

El ancho de la caja (Q3 − Q1) representa el **rango intercuartil** (IQR), que indica la **dispersión central**:
- **Caja angosta** → poca variabilidad central.
- **Caja ancha** → alta variabilidad central.

También se puede analizar la **dispersión general** observando el largo de los bigotes.

- **Posición de la mediana: simetría/asimetría**
- **Longitud de la caja: dispersión central (IQR)**
- **Longitud de los bigotes: dispersión general**
- **Puntos fuera de los bigotes: valores atípicos**

### 📊 Histograma vs. Diagrama de caja



Tanto el **histograma** como el **diagrama de caja** (boxplot) son herramientas útiles para explorar y visualizar conjuntos de datos.

Sin embargo, una **desventaja** del diagrama de caja es que puede mostrar **gráficos idénticos para conjuntos de datos muy diferentes**, si los cinco números resumen (mínimo, Q1, mediana, Q3 y máximo) son iguales o muy similares.

En el ejemplo de abajo vemos cuatro conjuntos de datos que tienen el mismo boxplot pero formas muy diferentes entre sí:
- uno puede tener una distribución uniforme,
- otro una asimetría marcada,
- otro muchos valores repetidos,
- y otro una concentración de datos en los extremos.

image.png

Esto **no se nota en el diagrama de caja**, pero sí se ve claramente en el **histograma** debajo.

📌 En resumen, tanto el histograma como el diagrama de cajas son útiles para ver la **forma de los datos** y la existencia de **datos atípicos** .

Sin embargo:
- El **histograma** es más útil para ver la **forma de los datos**.
- El **diagrama de caja** es más útil para **detectar datos atítipos**.
- En **ambos** puede observarse la **simetría** de los datos, aunque el histograma ofrece más detalle.
