### 🎯 ¿Qué es la población?

En estadística, la **población** se refiere al *conjunto completo* de unidades experimentales. Uno de los objetivos principales de Ciencia de Datos es dar respuesta a preguntas de interés en una población.

Por ejemplo, en nuestro problema de estudiar la densidad demográfica de los países del mundo, el conjunto de datos llamado `densidad_muestra_195` es **la población total de países del mundo** y no hay más que esos.

Sin embargo, en algunas situaciones, puede resultar imposible acceder a los datos de toda la población. Por ejemplo, en el caso de querer estudiar la resistencia al impacto de determinado vehículo. ¡No podríamos estrellar todos los vehículos fabricados para medir que tanto se rompen! Deberíamos dejar algunos para la venta.

Como esta, existen diferentes razones por las cuales no se puede acceder a la población:
- por imposibilidad técnica: población infinita (retomaremos este tema más adelante);
- porque la población puede ser muy dinámica y cambiante, siendo imposible estudiarla en su totalidad: pensemos en el ejemplo de todos los habitantes de una determinada ciudad, todos los días nacen y mueren personas;
- por razones económicas o temporales: población muy grande, ensayos destructivos, etc.


¿Pero entonces que hacemos en estos casos? ... Estudiamos muestras 🙂

### 🔍 ¿Qué es una muestra?

Una **muestra** es un subconjunto o parte de la población en que se llevará a cabo la investigación con el fin posterior de **generalizar** los hallazgos. A la cantidad de unidades experimentales que tomamos para formar una muestra le llamamos **tamaño muestral** y generalmente lo denotamos con $n$.

Por ejemplo, en nuestro problema de estudiar la densidad demográfica de los países del mundo, el conjunto de datos llamado `densidad_muestra_50` es **una muestra de tamaño 50 de la población total de países**, que son los 195 listados en el conjunto de datos llamado `densidad_poblacion_195`.

La rama de la estadística encargada de extraer conclusiones sobre la población entera a partir de la información extraída de una muestra se llama **inferencia estadística**, veremos más sobre esto en Ciencia de Datos II.

## 🧮 ¡A hacer cuentas!

Sigamos trabajando con los dos conjuntos de datos `densidad_muestra_50` y `densidad_poblacion_195`.

Empecemos calculando:
- La media de la **muestra** de tamaño 50: `densidad_muestra_50`.
- La media de la **población**: `densidad_poblacion_195`.

Para ello, podés usar esta función que calcula la media de una lista de números:

```python
def calcular_media(lista):
    return sum(lista) / len(lista)
```

Completá el código y calculá ambas medias:

```python
media_muestra_50 = calcular_media(densidad_muestra_50)
media_poblacional = calcular_media(densidad_poblacion_195)



print("Media de la muestra (n=50):", media_muestra_50)
print("Media de la población:", media_poblacional)
```

### ✅ Representatividad, sesgo y muestreo

Es muy importante que las muestras que se consideren para realizar cualquier análisis estadístico siempre sean **representativas** de la población, es decir, que posean las características básicas de toda la población y no de un subconjunto particular de la misma.

Esto nos sirve para, cuando podamos tener toda la población, confiar en que los resultados obtenidos en la muestra pueden extenderse a la población, y por lo tanto obtener conclusiones confiables para la misma.




Para lograr la representatividad, sería ideal seleccionar las unidades experimentales que
reflejen esas características básicas de toda la población. Sin embargo, al hacer esto, es posible que introduzcamos algún tipo de tipo de **sesgo** en el proceso de **muestreo**, lo que podría afectar la validez de nuestros resultados.

Para minimizar el sesgo, se recomienda elegir una **muestra aleatoria (simple)**. Esto quiere decir que cada unidad experimental debe ser seleccionada de manera **independiente**, **al azar** (y con la misma chance de salir que el resto de las unidades experimentales) de la población.

⛔ **¡Pero cuidado! La aleatoriedad por sí sola no garantiza que la muestra sea representativa de la población, ya que podemos tener la mala suerte de tomar al azar una muestra cuyas características no se parezcan a la de la población 🤯. Aún así, el muestreo aleatorio es el método más confiable para este fin.**
