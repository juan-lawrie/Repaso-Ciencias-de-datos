## 📝 Ejercicio 2

Ahora vamos a modificar la función `calcular_mediana` para que incluya un **parámetro opcional** que indique si la lista ya está ordenada o no.

El parámetro opcional será un **parámetro booleano con valor por defecto** **False**, llamado `ordenada`, para contemplar los siguientes casos:

- Si la lista **ya está ordenada**, la función debe crear una **copia** de la lista (para no modificarla por accidente dentro de la función).
- Si la lista **no está ordenada**, la función debe trabajar con una **nueva lista ordenada** sin modificar la original.

Completá el código para incorporar esta nueva funcionalidad:



```python
def calcular_mediana(lista, ordenada=False):
    if ordenada:
        lista_ordenada = lista.copy()
    else:
        lista_ordenada = sorted(lista)

    n = len(lista)
    mitad = n // 2

    if n % 2 == 0:
        mediana = (lista_ordenada[mitad-1] + lista_ordenada[mitad]) / 2
    else:
        mediana = lista_ordenada[mitad]

    return mediana
```

Ahora probá tu función con estos casos:

```python
# Lista NO ordenada (caso impar)
lista = [12, 7, 5, 15, 10]
print("Mediana:", calcular_mediana(lista))            # Mediana: 10
print("Lista original:", lista)                       # No debe cambiar
```

```python
# Lista NO ordenada (caso par)
lista = [12, 7, 5, 15]
print("Mediana:", calcular_mediana(lista))            # Mediana: 9.5
print("Lista original:", lista)                       # No debe cambiar
```

```python
# Lista YA ordenada (caso impar)
lista = [5, 7, 10, 12, 15]
print("Mediana:", calcular_mediana(lista,  ordenada=True))      # Mediana: 10
print("Lista original:", lista)                                 # No debe cambiar
```

```python
# Lista YA ordenada (caso par)
lista = [5, 7, 12, 15]
print("Mediana:", calcular_mediana(lista, ordenada=True))      # Mediana: 9.5
print("Lista original:", lista)                                # No debe cambiar
```

## 📝 Ejercicio 3

Ahora vamos a **reutilizar** la función `calcular_mediana`  para construir una nueva función que calcule los **cuartiles** de un conjunto de datos.

Recordá que los cuartiles dividen un conjunto de datos en cuatro partes iguales:
- **Q1 (primer cuartil)**: mediana de la mitad inferior (25% de los datos está por debajo).
- **Q2 (segundo cuartil)**: mediana del conjunto completo (50% de los datos está por debajo).
- **Q3 (tercer cuartil)**: mediana de la mitad superior (75% de los datos está por debajo).

Con esta información, completá la función para calcular y devolver los cuartiles:

```python
def calcular_cuartiles(lista):
    lista_ordenada = sorted(lista)

    n = len(lista_ordenada)
    mitad = n // 2

    if n % 2 == 0:
        mitad_inferior = lista_ordenada[:mitad]
        mitad_superior = lista_ordenada[mitad:]
    else:
        mitad_inferior = lista_ordenada[:mitad]
        mitad_superior = lista_ordenada[mitad + 1:]

    Q1 = calcular_mediana(mitad_inferior, ordenada=True)
    Q2 = calcular_mediana(lista_ordenada, ordenada=True)
    Q3 = calcular_mediana(mitad_superior, ordenada=True)

    return Q1, Q2, Q3
```

Ahora probá tu función calculando los cuartiles para estos datos:

```python
# Caso impar
lista = [15, 10, 12, 6, 9]

q1, q2, q3 = calcular_cuartiles(lista)  # Esperado: Q1=7.5, Q2=10, Q3=13.5

print("Q1:", q1)
print("Q2:", q2)
print("Q3:", q3)
```

```python
# Caso par
lista = [15, 10, 12, 6]

q1, q2, q3 = calcular_cuartiles(lista)  # Esperado: Q1=8, Q2=11, Q3=13.5

print("Q1:", q1)
print("Q2:", q2)
print("Q3:", q3)
```

A simple vista, algunos valores parecen muy bajos o muy altos comparados con el resto.

Para saber si son **atípicos**, vamos a usar la **regla del IQR** (rango intercuartíl) donde un valor se considera atípico si:

- es **menor** que $Q1 − 1.5 \times IQR$, **o**
- es **mayor** que $Q3 + 1.5 \times IQR$

Completá la función `detectar_atipicos` para calcule y retorne una lista con los **valores considerados atípicos según la regla del IQR**.

Para calcular los cuartiles reutilizá la función `calcular_cuartiles` que definiste antes.

```python
def detectar_atipicos(lista):
    Q1, Q2, Q3 = calcular_cuartiles(lista)

    IQR = Q3 - Q1
    LI = Q1 - 1.5 * IQR
    LS = Q3 + 1.5 * IQR

    atipicos = []

    for dato in lista:
        if dato < LI or dato > LS:
            atipicos.append(dato)

    return atipicos
```
